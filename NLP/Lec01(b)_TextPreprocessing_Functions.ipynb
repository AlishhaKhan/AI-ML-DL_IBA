```python
# 1. Change text into lowercase
def ChangeCase(text):
    return text.lower()
```

```python
# 2. Remove punctuation and special characters
def RmovePunctuationSpChar(text):
    punctuation_pattern = r'[^\w\s]'
    cleaned_text = re.sub(punctuation_pattern, '', text)
    return cleaned_text
```

```python
# 3. Remove URLs
def RemoveURL(text):
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    return url_pattern.sub(r'', text)
```

```python
# 4. Remove HTML tags
def RemoveHTML(text):
    html_tags_pattern = r'<.*?>'
    text_without_html = re.sub(html_tags_pattern, '', text)
    return text_without_html
```

```python
# 5. Function to remove stopwords
def RemoveStopwords(txt, lang):
    stop_words = set(stopwords.words(lang))   # get stopwords of selected language
    txt = txt.lower()                         # convert sentence into lowercase
    word_tokens = txt.split()                # split sentence into individual words
    filtered_text = [word for word in word_tokens if word not in stop_words]
    # keep only those words which are not stopwords
    return filtered_text
```

```python
# Function to remove stopwords
def RemoveStopwords(txt, lang):
    stop_words = set(stopwords.words(lang))   # get stopwords list of selected language
    txt = txt.lower()                         # convert text into lowercase
    word_tokens = nltk.word_tokenize(txt)    # tokenize sentence into words
    filtered_text = [word for word in word_tokens if word not in stop_words]
    return filtered_text
```
