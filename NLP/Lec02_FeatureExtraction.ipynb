```python
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
stopwords = set(stopwords.words("english"))
```

```python
sample_text = "this is a sample text"
tokens = sample_text.split()
clean_tokens = [t for t in tokens if not t in stopwords]
clean_text = " ".join(clean_tokens)
print(clean_tokens)
print(clean_text)
```

**Porterstemmer**

```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
```

```python
words = ['running', "flies", "happily", "better", "jumps", "happiest"]
```

```python
stemmed_words = [stemmer.stem(word) for word in words]
print(stemmed_words)
```

**Lancaster Stemmer**

```python
from nltk.stem import LancasterStemmer
stemmer = LancasterStemmer()
```

```python
stemmed_words = [stemmer.stem(word) for word in words]
print(stemmed_words)
```

**Snowball Stemmer**

```python
from nltk.stem import SnowballStemmer
stemmer = SnowballStemmer("english")
```

```python
stemmed_words = [stemmer.stem(word) for word in words]
print(stemmed_words)
```

**RegExp Stemmer**

```python
from nltk.stem import RegexpStemmer
stemmer = RegexpStemmer("ing$|es$|ly$|er$", min=4)
```

```python
stemmed_words = [stemmer.stem(word) for word in words]
print(stemmed_words)
```

**WordNet Lemmatizer**

```python
from nltk.stem import WordNetLemmatizer
nltk.download("wordnet")
lemmatizer = WordNetLemmatizer()
```

```python
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in words]
print(lemmatized_words)
```

**Part of Speech Tagging**

```python
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger_eng')
nltk.download('tagsets_json')
```

```python
sentence = "the quick brown fox jumpy over the lazy dog"
words = nltk.word_tokenize(sentence)
pos_tags = nltk.pos_tag(words)

print("")
print("POS Tags\n", pos_tags, "\n")

for (word, tag) in pos_tags:
    print(nltk.help.upenn_tagset(tagpattern=tag))
```

```python
from nltk.tree import Tree
```

```python
sentence = "The big brown dog barked loudly."
words = nltk.word_tokenize(sentence)
pos_tags = nltk.pos_tag(words)

chunked = nltk.ne_chunk(pos_tags)

reg_exp = "NP: {<DT>?<JJ>*<NN>*}"
rp = nltk.RegexpParser(reg_exp)
result = rp.parse(chunked)

result
```
