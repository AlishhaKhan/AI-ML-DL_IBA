**📘 PARSING TECHNIQUES**

**1. What is Parsing?**

Parsing is the process of analyzing the grammatical structure of a sentence.

Easy Understanding:

Machine checks:

which word is noun

which word is verb

which words are connected

Example:

```python
"The dog barked loudly"
```

Machine understands:

dog → noun

barked → action

loudly → describes action

**Why Parsing Important?**

Because machine must understand “sentence ka structure”
not just words.

**2. TYPES OF PARSING**

**POS TAGGING (Part of Speech Tagging)**

Assigning grammatical role to each word.

Example:

| Word   | Tag |
| ------ | --- |
| The    | DT  |
| dog    | NN  |
| barked | VBD |
| loudly | RB  |

**Common POS Tags**

| Tag | Meaning    |
| --- | ---------- |
| NN  | Noun       |
| VB  | Verb       |
| JJ  | Adjective  |
| RB  | Adverb     |
| DT  | Determiner |

**Tools & Techniques**

Using:

NLTK



```python
nltk.pos_tag()
```

**Types of POS Tagging**

**1. Rule-Based POS Tagging**

Uses grammar rules.

Example: words ending with “ing” → probably verb

**2. Statistical POS Tagging**

Uses probability and datasets. More accurate.

**3. Hybrid POS Tagging**

To identify: nouns
verbs
sentence structure

Used in: chatbots
grammar checkers
translation systems

**3. CHUNKING (Shallow Parsing)**

Grouping words into meaningful phrases.

```python
"The big brown dog"
```

becomes:

```python
(NP The big brown dog)
```

NP = Noun Phrase

**Why “Shallow Parsing”?**

Because: it identifies phrases
but not complete deep sentence structure

**Chunking Rules**

Example Rule:

```python
NP: {<DT>?<JJ>*<NN>*}
```

**Meaning of Rule**

| Symbol | Meaning      |
| ------ | ------------ |
| DT     | determiner   |
| ?      | optional     |
| JJ     | adjective    |
| *      | zero or more |
| NN     | noun         |

**Simple Understanding:**

If sentence has:

article
adjectives
noun

→ make noun phrase.

**Chunk Tags**

| Tag | Meaning              |
| --- | -------------------- |
| NP  | Noun Phrase          |
| VP  | Verb Phrase          |
| PP  | Prepositional Phrase |

**Tools & Techniques**

```python
nltk.RegexpParser()
```

**Why Chunking Used?**

To identify:

meaningful groups, phrases, sentence sections

**4. CONSTITUENCY PARSING**

Breaking sentence into sub-phrases using tree structure.

Example:

```python
(S
   (NP The dog)
   (VP barked loudly)
)
```

**Understanding**

Sentence divided into noun phrase and verb phrase.

**Why Used?**

To understand complete sentence grammar & hierarchical structure.

**5. DEPENDENCY PARSING**

Shows relationship between words in a sentence.

Example:

```python
"The dog barked loudly"
```

Relations:

dog → subject

barked → main verb

loudly → modifier

**Simple Understanding**

Dependency parsing focuses on “which word depends on which word”.

**Difference Between Constituency & Dependency Parsing**

| Constituency Parsing | Dependency Parsing    |
| -------------------- | --------------------- |
| phrase-based         | relation-based        |
| tree of phrases      | word relationships    |
| focuses on structure | focuses on dependency |


**📘 FEATURE EXTRACTION**

**1. What is Feature Extraction?**

Converting text into numerical format for machine learning.

**Why Needed?**

Because machine understands numbers, not text.

**FEATURE EXTRACTION TECHNIQUES**

**2. BAG OF WORDS (BOW)**

Represents text using word frequency.

**Process of BOW**

Step 1:

Collect documents.

Step 2:

Tokenization

Sentence → words

Step 3:

Create vocabulary.

Step 4:

Count word frequencies.

Example:

```python
"I love AI"
"I love ML"
```

Vocabulary:

```python
[I, love, AI, ML]
```

Vectors:

| Doc | I | love | AI | ML |
| --- | - | ---- | -- | -- |
| D1  | 1 | 1    | 1  | 0  |
| D2  | 1 | 1    | 0  | 1  |

**Numeric Vectors in BOW**

Each sentence becomes:

```python
[1,1,1,0]
```

This is called vector representation.

**Vector Difference**

Different sentences produce different vectors. Because each sentence has different words/counts.

**3. BINARY BAG OF WORDS**

Checks only presence or absence of words.

Example:

If word exists: 1, if not: 0

**Difference**

| BOW              | Binary BOW           |
| ---------------- | -------------------- |
| counts frequency | checks presence only |

**LIMITATIONS OF BOW**

ignores word order

ignores meaning

large sparse matrix

no context understanding

**4. TF-IDF**

Measures importance of a word in document collection.

**TF (Term Frequency)**

Checks word kitni dafa document mein aaya

**IDF (Inverse Document Frequency)**

Checks word kitne documents mein common hai

**Final Idea**

common words → low importance

rare important words → high importance

**Why TF-IDF Better Than BOW?**

Because it gives weighted importance and removes effect of very common words.

Example: the - appears everywhere → low score, AI - more meaningful → higher score

**CASE STUDY (Simple Understanding)**

Suppose bank has customer reviews.

```python
"bank service is terrible"
```

TF-IDF identifies:

“terrible” important
“is” not important

Then ML model predicts:
→ negative review
