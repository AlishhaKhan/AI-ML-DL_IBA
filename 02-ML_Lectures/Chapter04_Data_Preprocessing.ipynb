**PART B — MISSING VALUES IMPUTATION**

**1. WHAT ARE MISSING VALUES?**

Dataset mein blank/unknown/null entries.

Represent: NaN, Null, ?, empty cell

Reasons: user forgot input, sensor failed, survey incomplete, data corruption

**2. TYPES OF MISSING DATA**

Generally 3 conceptual types:

**MCAR — Missing Completely At Random**

randomly missing, no pattern

Example: internet issue se kuch forms lost.

**MAR — Missing At Random**

depends on some other observed variable.

Example: older people salary disclose kam kar rahe.

**MNAR — Missing Not At Random**

missing due to hidden reason linked with itself

Example: very high earners salary intentionally skip

Conceptually enough.

**3. WHY MISSING VALUES ARE DANGEROUS?**

Because: average distort hota, algorithm errors, pattern incomplete, bias create

Many sklearn models NaN accept hi nahi karte, so we must handle them.

**4. LISTWISE DELETION**

Simple method: rows with missing values delete.

Example:

| Age | Salary |
| --- | ------ |
| 22  | 30000  |
| NaN | 25000  |

delete second row

Advantage: easy,
Disadvantage: data loss

Useful only when few rows missing.

**5. SIMPLE IMPUTATION — MEAN / MODE**

Instead of deleting, fill.

**Numerical column → Mean**

Ages: 20, 25, 30, NaN

Mean = 25

Fill NaN with 25

**Categorical column → Mode**

City: Karachi, Lahore, Karachi, NaN

Mode = Karachi

This preserves rows.

**6. CONSTANT VALUE IMPUTATION**

Fill with fixed value

Example: salary missing → 0 or "Unknown"

Used when you want explicit placeholder.

**7. FORWARD FILL / BACKWARD FILL**

Mostly time-series style.

Data: 10, NaN, NaN, 15

**Forward fill:**

Forward fill: 10, 10, 10, 15

*Previous value carry*

**Backward fill:**

10, 15, 15, 15

*Next value carry*

Useful when sequential continuity expected.

**8. LINEAR INTERPOLATION**

Estimate missing value between two points using trend.

10, NaN, 20

middle assumed = 15

Because linearly between 10 and 20. Useful for continuous sequential numeric data.

**9. KNN IMPUTATION**

Advanced and smart, Machine looks at nearest similar rows and fills missing based on neighbors.

Example: similar age/salary users dekh kar missing income estimate, more accurate than mean sometimes, but computationally heavier.

**⭐ SUMMARY OF IMPUTATION METHODS**

delete row, mean/mode, constant, forward/backward, interpolation, KNN

Choice depends on data nature.

**PART C — ENCODING (Categorical to Numeric)**

This is SUPER IMPORTANT, Machine learning algorithms mostly numbers samajhte.

Text labels: Karachi, Red, Yes/No, Machine cannot do algebra on words.

So encoding = converting categorical text into numeric representation.

**CATEGORICAL DATA**

Non-numeric labels/categories.

Examples: gender, city, grade, yes/no, Need conversion.

**1. LABEL ENCODING**

Each category gets integer.

Yes = 1 No = 0;
Karachi = 0 Lahore = 1 Islamabad = 2

Problem: the machine may think 2 >1 >0, meaning order, when actually cities have no ranking, so careful.

**2. ONE HOT ENCODING**

Best for nominal categories.

City column becomes:

| Karachi | Lahore | Islamabad |
| ------- | ------ | --------- |
| 1       | 0      | 0         |
| 0       | 1      | 0         |


Each category separate binary column, no fake ranking, widely used.

**3. BINARY ENCODING**

Categories first integer then binary form, useful when categories bohot zyada hon and one-hot huge ho jaye, Compact storage.

**4. TARGET ENCODING**

Each category replaced by target mean.

Example: City wise average purchase rate.

Karachi = 0.8, Lahore = 0.3

Useful and advanced but leakage ka risk hota.

**5. ORDINAL ENCODING**

When categories have natural order.

Low, Medium, High

Low = 1 Medium = 2 High = 3

Yahan ranking valid hai.

**6. CYCLIC ENCODING**

For circular values: months, days, hours

Because month 12 aur month1 actually close hote.

Simple numeric 12 and1 misleading.

So sine/cosine transform use hota.

If: blanks, text, inconsistent values, then model confused.

Preprocessing gives: clean, complete, numeric, structured data.

Thus: better learning.
