**🟢 CHAPTER 9 — UNSUPERVISED LEARNING (K-MEANS CLUSTERING)**

*(Grouping Similar Data Without Teacher)*

**1. WHAT IS UNSUPERVISED LEARNING?**

Remember chapter 1: Supervised = labeled data (teacher present)

Unsupervised = unlabeled data

Meaning: machine ko sirf input data diya jata hai, correct answers nahi diye jate.

No target column, Machine khud hidden patterns ya groups discover karti hai.

Example: Suppose customer dataset:

| Age | Income | Spending |
| --- | ------ | -------- |

No label saying rich/poor or premium/normal.

Machine khud dekhegi:
“acha ye customers similar lag rahe”, “ye alag cluster”

This self grouping = unsupervised learning.

**2. MAIN TYPES OF UNSUPERVISED LEARNING**

Clustering, Association, Dimensionality Reduction

But in syllabus we have = clustering

**3. WHAT IS CLUSTERING?**

Clustering means grouping similar data points together.

Like making friend circles from class students based on similarity.

Same nature wale ek group, Different nature wale doosra, No teacher tells groups. Machine forms groups itself.

**4. WHAT IS K-MEANS CLUSTERING?**

K-Means is an unsupervised algorithm that partitions data into K number of clusters based on similarity.

K means number of groups we want.

Example: K=3 means machine data ko 3 groups mein divide kare.

**5. WHY CALLED K-MEANS?**

K = chosen number of clusters

Means = each cluster represented by its mean center point.

That center point = centroid.

**6. CENTROID**

Centroid means central average point of a cluster.

Think: har group ka middle representative location.

Data points nearest to same centroid belong to same cluster. Centroid acts like cluster leader.

**7. HOW K-MEANS WORKS? (CORE MOVIE)**

Algorithm ka pura ka pura kaam repeated shifting hai.

**Step 1:**

Choose K value, Suppose K=2, Need 2 clusters.

**Step 2:**

Randomly place 2 centroids, Initial guesses.

**Step 3:**

Assign each data point to nearest centroid.

Means: jo point centroid 1 ke paas us cluster mein,
jo centroid 2 ke paas us mein, un dono points ko hum assign kar dien ge.

**Step 4:**

Now recalculate centroids, Each formed cluster ka new average center, Centroid shifts.

**Step 5:**

Again reassign points to nearest new centroid, Some points cluster change karenge.

**Step 6:**

Repeat centroid shifting + reassignment, until centroids stop changing significantly. Then clusters stabilized.

**8. K-MEANS STEPS IN SHORT**

Choose K

↓

initialize centroids

↓

assign nearest points

↓

recompute centroids

↓

repeat until stable

This loop is the entire chapter.

**9. SIMPLE VISUAL EXAMPLE**

Suppose mall customers:

some low spend low income

some high income high spend

some average

K=3 choose.

Algorithm naturally creates 3 customer groups, Now business can target each group differently.

**10. WHY K-MEANS USEFUL?**

Because many times hume labels pata hi nahi hote. Need hidden segmentation.

Examples: customer segmentation, image compression, document grouping, social media audience groups, anomaly pre-analysis

**11. PROS OF K-MEANS**

simple,
fast,
easy implement,
works well on clear clusters,
scalable

**12. CONS OF K-MEANS**

K pehle choose karna padta,
random centroid se result vary,
outliers affect,
non-spherical weird clusters mein weak

**13. PRACTICAL APPLICATIONS**

marketing segmentation

recommendation grouping

medical patient categories

pattern discovery
