**PART A — WHY MODEL EVALUATION IS NEEDED?**

Suppose machine ne predictions de diye. But kitne sahi?, kitne ghalat?, kis type ki ghalti?, numeric error kitna?, classification reliable hai?

Without checking this, trained model useless hai. So evaluation means measuring model performance scientifically.

**PART B — TWO EVALUATION WORLDS**

Remember chapter 5: there were two supervised tasks: classification & regression

So naturally evaluation bhi 2 types: Classification Metrics (for yes/no class models) & Regression Metrics (for numeric models)

Regression metrics hum chapter 6 mein kar chuke: MSE, MAE, RMSE, MAPE, R²

So yahan classification deep.

**PART C — CONFUSION MATRIX (MAIN HERO)**

Confusion Matrix is a table that compares actual classes with predicted classes.

Suppose binary classification: Pass / Fail

Matrix counts 4 situations.

**1. TRUE POSITIVE (TP)**

Actually positive and predicted positive.

Example: student pass tha, model ne pass bola. Correct!

**2. TRUE NEGATIVE (TN)**

Actually negative and predicted negative. Fail tha, fail bola. Correct!

**3. FALSE POSITIVE (FP)**

Actually negative but predicted positive. Fail tha, model ne pass bol diya. Wrong positive!

**4. FALSE NEGATIVE (FN)**

Actually positive but predicted negative. Pass tha, model ne fail bol diya. Wrong negative!

**Why called confusion matrix?**

Because it shows machine exactly kis jagah confuse hui.

Not just total marks.

**PART D — ACCURACY**

Accuracy means total correct predictions / total predictions.

Accuracy = TP+TN/TP+TN+FP+FN

Simple: overall kitna percent sahi. If 90 out of100 correct =90%.

**BUT ACCURACY CAN MISLEAD**

Suppose 95 students pass, 5 fail. Model har kisi ko pass bol de. Accuracy =95%. Looks amazing. But fail students kabhi identify nahi hue. So need deeper metrics.

**PART E — PRECISION**

Precision asks among all predicted positives, kitne actually positive the?

Precision = TP/TP+FP

Meaning: jab model “Yes” bole, kitna trustworthy? Useful when false alarms costly.

**PART F — RECALL**

Recall asks among all actual positives, kitne model ne catch kiye?

Recall = TP/TP+FN

Meaning: real positives miss to nahi ho rahe?

Useful when missing dangerous.

Example disease detection.

**PART G — PRECISION vs RECALL DIFFERENCE**

Precision: model ke YES predictions kitne pure?

Recall: real YES cases kitne pakre?

One focuses false positives, other false negatives.

**PART H — F1 SCORE**

Sometimes precision high, recall low. Need balanced combined score.

F1 = 2X Precision × Recall/Precision + Recall

Higher F1 = balanced strong classifier.

**PART I — FINAL QUICK RECAP OF REGRESSION METRICS**

Already done but connect here:

**MAE:**

average absolute miss

**MSE:**

average squared miss

**RMSE:**

sqrt miss

**R²:**

explained variance

These used when target numeric.

**PART J — WHEN TO USE WHICH METRIC?**

**Balanced simple dataset:**

Accuracy okay

**False positive concern:**

Precision

**False negative concern:**

Recall

**Need balanced:**

F1

**Numeric prediction:**

MSE/MAE/RMSE/R²
