**PART A — STATISTICS INTRODUCTION**

**1. WHAT IS STATISTICS?**

Statistics is the science of collecting, organizing, analyzing and interpreting data.

Simple: Data ko samajhne ka ilm = statistics

Machine learning works on data, so before machine learns, that human understanding comes from statistics.

human ko data samajhna padta hai.

**2. STATISTICAL ANALYSIS**

Statistical analysis means: applying mathematical methods to data to extract useful information.

Like: average kya hai?,
spread kitna hai?,
trend hai?,
unusual values hain?,
relation hai?

Before ML model: we analyze dataset statistically.

Why? Because: because without understand data we can't train model.

**3. WHAT IS STATISTICAL DATA?**

Data = observations/information

Example student data:

| Name | Hours Study | Marks |
| ---- | ----------- | ----- |

These rows are statistical data points.

Each row = observation

Each column = variable.

**4. TYPES OF VARIABLES**

This is very important.

**(A) Qualitative / Categorical Variable**

Names/categories

Examples: gender, color, city, pass/fail

These are labels not measured numbers.

**(B) Quantitative / Numerical Variable**

Numbers

Examples: age, salary, marks, weight

Can be measured mathematically.

**Further numerical two types:**

**Continuous** any real value (height, temperature)

This distinction will be use in distribution.

**PART B — DESCRIPTIVE STATISTICS**

Descriptive statistics means: summarizing data in short understandable numbers.

Instead of looking 1000 rows, we calculate some summary.

Main tools:

**1. CENTRAL TENDENCY**

Central tendency means: data ka center ya typical value.

Three common:

Mean = average; sum / total observations

Median = middle value

Mode = most frequent value

Example: 2, 3, 3, 5, 10

Mean = 4.6

Median = 3

Mode = 3

Each tells center differently.

**2. WHY AVERAGE ALONE IS NOT ENOUGH?**

Suppose: Dataset A = 4,5,6; Dataset B = 1,5,9

Both mean = 5

But clearly second dataset zyada spread hai, So we need spread measures.

**3. VARIANCE**

Variance tells: data average se kitna dispersed hai.

Formula concept: each value - mean, square it, average it

Bigger variance = values far apart

Smaller variance = values close

Variance = ∑(x−xˉ)2​/n

No need ratta now, understand, it measures inconsistency.

**4. STANDARD DEVIATION**

Variance ka square root:
σ = ∑(x−xˉ)2/n​

Why needed? Because variance squared units mein hoti, Standard deviation original units mein spread batata.

Simple: higher SD = more scattered data

Example: Marks: 48,50,52 = low spread

Marks: 10,50,90 = high spread

Average similar ho sakta, but SD different.

**5. COEFFICIENT OF VARIATION (CV)**

Sometimes two datasets have different units ke hote, so they need a relative spread.

CV = SD / Mean × 100

CV= xˉσ × 100

Lower CV = more consistency, Used for comparing variability.

**6. QUANTILES**

Quantiles divide ordered data into equal parts.

Examples: Quartiles = 4 parts; Percentiles =100 parts

Important:

Q1 = 25%

Q2 = median

Q3 = 75%

Helps understand data distribution.

**7. IQR (Interquartile Range)**

Middle 50% spread

IQR = Q3​ − Q1

Useful for detecting outliers

Small IQR = middle data compact.

**8. BOXPLOT IDEA**

Boxplot visually shows:

min

Q1

median

Q3

max

outliers

This gives data spread snapshot.

**PART C — PROBABILITY FOUNDATION**

**1. WHAT IS PROBABILITY?**

Probability means: chance of an event happening.

Value between 0 and 1

0 = impossible
1 = certain

Example:
coin head = 0.5

Machine learning uses probability to handle uncertainty.

**2. COMPUTING PROBABILITY**

Basic:
P(E) = Favorable Outcomes​/Total Outcomes

Dice pe even number: 3 favorable / 6 total = 0.5

**3. BAYES THEOREM**

This is super important ML theorem.

Bayes helps: update probability after getting new evidence.

Simple meaning: if symptom mila, disease ki probability update karo.

Spam filters also use Bayes.

**PART D — PROBABILITY DISTRIBUTIONS**

**1. WHAT IS A DISTRIBUTION?**

Distribution tells: data/probability values kis pattern mein spread hain.

Not every number equally likely, Some center ke paas zyada hote, Some rare.

This pattern = distribution.

**2. TYPES OF DISTRIBUTION**

**DISCRETE DISTRIBUTION**

count values example dice outcomes

**BINOMIAL DISTRIBUTION**

only two outcomes repeated:
success/failure

Example:
10 students pass/fail.

Used when:
fixed trials + two outcomes

**CONTINUOUS DISTRIBUTION**

real values possible:
Height, weight, temperature

**NORMAL DISTRIBUTION**

Most famous bell shape, center pe zyada values, extremes pe kam

Human height, exam marks often approx normal.

Why important? Because ML many assumptions use near-normal behavior

**STANDARD NORMAL DISTRIBUTION**

Normal distribution standardized: mean = 0; sd =1

This helps universal comparison.

**3. Z SCORE TRANSFORMATION**

How far a value from mean in SD units.

z=2 means:

value is 2 standard deviations above mean

Useful in anomaly detection and inference.

**PART E — CORRELATION + SHAPE OF DATA**

**1. CORRELATION INTRO**

Correlation means: strength and direction of relationship between two variables.

Example: study hours ↑ marks ↑

positive relation

**2. CORRELATION COEFFICIENT**

Range: -1 to +1

+1 = strong positive,
-1 = strong negative,
0 = no relation

Closer to magnitude 1 = stronger relationship.

**3. SKEWNESS**

Skewness = asymmetry

Data one side longer tail

Right skew / left skew

Shows data imbalance

**4. KURTOSIS**

Kurtosis = peakness/heavy tails

Shows whether distribution has many extreme values.

**5. RELATION WITH VARIANCE**

Variance spread batata, Skewness shape tilt batata, Kurtosis tail/extreme behavior batata hai, Together they fully describe data behavior.
