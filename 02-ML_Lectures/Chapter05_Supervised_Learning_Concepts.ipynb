**🟢 CHAPTER 5 — SUPERVISED LEARNING CORE CONCEPT**

*(Teacher-Guided Machine Learning)*

**1. WHAT IS SUPERVISED LEARNING?**

Supervised learning is a type of machine learning in which the model is trained on labeled data.

Labeled data means input ke saath correct answer bhi diya hota hai.

Teacher present.

Example:

| Study Hours | Marks |
| ----------- | ----- |
| 2           | 40    |
| 5           | 80    |

Here machine sees: hours + correct marks.

Then relation learns, So supervised = guided learning.

**2. WHY CALLED “SUPERVISED”?**

Because supervision = monitoring by teacher

Human analogy: teacher student ko questions + correct answers de.

Student pattern samajh jata, machine bhi same, and data acts like teacher, Hence supervised.

**3. HOW SUPERVISED MODEL LEARNS?**

Machine repeatedly does:

Input dekho

↓

Output dekho

↓

guess relation

↓

error check

↓

improve relation

Thousands of examples ke baad:

it can predict unseen outputs.

This is teacher-led pattern extraction.

**4. TWO STEP PROCESS OF SUPERVISED LEARNING**

**STEP 1 — TRAINING**

Labeled historical data feed, Machine learns mapping.

X → Y

Input to Output relation.

**STEP 2 — PREDICTION/TESTING**

Now new unseen X do, Machine predicts Y.

If prediction close/correct = learning successful.

Simple but this is whole supervised concept.

**5. MAIN TYPES OF SUPERVISED LEARNING**

There are 2 major tasks:

**(A) Classification**

**(B) Regression**

This difference is EXTREMELY IMPORTANT. Every upcoming algorithm in one of these purposes used.

**PART A — CLASSIFICATION**

**WHAT IS CLASSIFICATION?**

Classification means predicting a category/class label.

Output is discrete class.

Examples: Yes/No, Spam/Not Spam, Pass/Fail, Disease/No Disease, Not continuous numbers. Machine chooses class.

**CLASSIFICATION CASE STUDY**

Suppose college wants predict whether student will pass.

Dataset:

| Attendance | Hours Study | Pass |
| ---------- | ----------- | ---- |
| 90         | 5           | Yes  |
| 40         | 1           | No   |
| 80         | 4           | Yes  |

Model sees many examples, Learns: high attendance + study = mostly yes

New student: 70 attendance, 3 hours Model predicts: Pass = Yes maybe

This is classification because answer category hai.

**CLASSIFICATION REAL EXAMPLES**

Email spam filter, Fraud / not fraud, Tumor benign/malignant, Customer churn yes/no, Face recognized/not recognized. All are class decisions.

**PART B — REGRESSION**

**WHAT IS REGRESSION?**

Regression means predicting a continuous numeric value.

Output category nahi — exact number

Examples: house price, salary, marks, temperature, sales amount, Machine estimates quantity.

**REGRESSION CASE STUDY**

Dataset:

| House Size | Rooms | Price   |
| ---------- | ----- | ------- |
| 1200       | 3     | 50 lakh |
| 1800       | 4     | 80 lakh |

Machine learns numeric relationship.

New house: 1500 sqft, 3 rooms; Predict: 65 lakh

This is regression because output numerical.

**CLASSIFICATION vs REGRESSION (MOST IMPORTANT)**

| Classification   | Regression       |
| ---------------- | ---------------- |
| Predict category | Predict number   |
| Yes/No labels    | Continuous value |
| Pass/Fail        | Marks prediction |
| Spam/Not spam    | House price      |

One line:

if output group name → classification

if output measurable quantity → regression

**6. HOW SUPERVISED LEARNING FITS INTO PIPELINE?**

Remember chapter 2, choose model step, at that point:

if target categorical → choose classification algorithm

if target numeric → choose regression algorithm

So supervised learning is not one algorithm, It is a family containing many algorithms.
