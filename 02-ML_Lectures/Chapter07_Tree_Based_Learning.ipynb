**PART A — DECISION TREE**

**1. WHAT IS DECISION TREE?**

Decision Tree is a supervised learning algorithm that makes predictions by asking a sequence of questions and splitting data into branches like a tree.

Simple: machine yes/no type questions pooch kar final decision tak pohanchti hai. Exactly human reasoning tree.

Example: Should customer buy product?

Question 1: Age > 30?

if yes → ask income?

if no → ask student?

Eventually final answer = Buy / Not Buy

This branching structure = decision tree.

**2. WHY CALLED TREE?**

Because diagram looks like upside-down tree:

top = root, middle = branches/nodes, bottom = leaves

Decision grows downward.

**3. DATA → TARGET CONCEPT**

Like every supervised model:

Input features:
age, income, student status

Target:
buy/not buy

Decision tree tries to create rules from data that best lead to target.

**4. DECISION PROCESS (HOW TREE THINKS?)**

Machine asks: which feature should I split on first so data most clearly separate ho jaye?

Then: within each subgroup, next best question.

Then next, Until final class mostly pure.

So tree = repeated splitting.

**5. ROOT NODE**

Topmost first decision point, Most important split.

Example: "Age > 30?" may be root

Root node contains full dataset initially.

**6. BRANCHES**

Outcome paths from a node.

Yes branch / No branch.

Each branch carries subset of data.

**7. LEAF NODE / FINAL DECISION**

Bottom final prediction.

Example: Buy = Yes

No further questions.

**8. DECISION RULES**

Every path forms IF-THEN rule.

IF age>30 AND income high THEN buy=yes

So tree can be translated into human-readable rules.

This is why DT is interpretable.

**9. ENTROPY (VERY IMPORTANT)**

Now main mathematical question: How decide best split?

Using impurity measure.

Entropy means amount of disorder/uncertainty in data.

If node has mixed yes/no heavily → entropy high.

If node mostly one class → entropy low.

Formula: Entropy = − ∑pi log2 pi

No need derivation, Meaning only, machine wants splits that reduce confusion.

**10. BEST SPLIT IDEA**

Machine checks: if I split on Age, entropy kitni reduce?

if split on Income, entropy kitni reduce?

Whichever gives maximum purity = choose.

This reduction often called information gain.

**11. CLASSIFICATION TREE vs REGRESSION TREE**

**Classification Tree**

final output category.

Buy/Not Buy

**Regression Tree**

final output number.

House price etc.

Same tree logic, different target type.

**12. TYPES COMPARISON (classification/regression)**

Classification: leaf = class label

Regression: leaf = numeric average/value

**13. OVERFITTING IN DECISION TREE**

Very common issue, Tree keeps splitting too much,
memorizes every tiny detail.

Training perfect, new data poor, This is overfitting. Tree becomes too deep and specific.

**14. TREE PRUNING**

Pruning means cutting unnecessary branches.

Simplify tree, Remove overly detailed weak splits.

Result: better generalization. Like plant ki extra shaakhein kaatna.

**15. DT INDUCTION**

Decision tree induction means process of constructing the tree from training data using recursive splitting.

Recursive = split, then split subsets, then split subsets...

Until stopping condition. This whole build process = induction.

**16. DECISION TREE ALGORITHM SUMMARY MOVIE**

Full data at root

↓

choose best split by entropy

↓

make branches

↓

repeat recursively

↓

reach leaves

↓

prune if too large

Done.

**PART B — RANDOM FOREST**

Now think: single decision tree easy hai. But one tree can be unstable. Small data change = tree changes. Need stronger prediction.

So idea: instead of one tree, make many trees and combine. That is Random Forest.

**17. WHAT IS RANDOM FOREST?**

Random Forest is an ensemble supervised learning method that builds multiple decision trees and combines their outputs for better accuracy.

Forest = many trees.

**18. ENSEMBLE METHODS**

Ensemble means many weak/base models combine to make strong model.

Like class ka collective vote.

Random forest = ensemble of trees.

**19. MAIN TYPES OF ENSEMBLE (overview)**

Bagging and Boosting

Random forest mainly bagging based.

**20. BOOTSTRAP SAMPLING**

This is first magic. Instead of giving same exact data to each tree, random forest creates random sampled datasets (with replacement) called bootstrap samples.

Means: Tree1 sees slightly different rows, Tree2 different rows, Tree3 different rows. This diversity important.

**21. TREE CONSTRUCTION**

Now each sampled dataset pe separate decision tree build hota.

But not just row randomness— feature randomness too.

**22. RANDOM FEATURE SELECTION**

At every split, each tree does not check all features.

It checks random subset of features, This makes trees less similar.

Hence "random."

**23. SPLITTING CRITERIA / POTENTIAL SPLIT POINT / IMPURITY REDUCTION**

Each tree internally still uses impurity logic: entropy/gini etc.

Chooses split giving highest impurity reduction.Same DT principle inside each tree.

Potential split point = possible threshold options.

Best one selected.

**24. HIGHEST IMPURITY REDUCTION**

Whichever question makes node most pure = selected.

Again repeated recursively.

**25. DECISION TREE FOREST**

After many trees built: we now have forest of predictions.

Tree1 says Yes

Tree2 says Yes

Tree3 says No

Tree4 says Yes

Final answer by majority vote (classification)
or average (regression)

This gives robust output.

**26. FINAL PREDICTION**

**Classification RF:** majority voting

**Regression RF:** mean average of tree outputs

**27. OOB EVALUATION (OUT OF BAG)**

Because bootstrap sample with replacement means some rows left out for each tree. Those unused rows = Out Of Bag samples.

Can use them as mini validation to estimate performance without separate test.Smart internal checking.

**28. PROS OF RANDOM FOREST**

more accurate than single DT

less overfitting

robust

handles nonlinear patterns

good on many datasets

**29. CONS OF RANDOM FOREST**

slower

less interpretable than one tree

many trees = computational load
