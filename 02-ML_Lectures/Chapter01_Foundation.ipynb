**2. WHAT IS LEARNING? (before ML)**

Question: How does a machine learn?

Humans learn from mistakes and experiences.

Machine learning with the same philosophy, learning patterns from past data and improve them for future decisions.

Meaning: We does not tell rules manually to the machine. We just give examples. Then the machine learns by itself.

Example: 1000 emails do: spam / not spam

Machine notice pattern: "free", "offer", "win money" etc.

Then new email pop up: machine says = it's probably spam.

This improvement from experience = learning.

**3. WHAT IS MACHINE LEARNING?**

Machine Learning is a subset of AI in which computers learn patterns from data and make predictions/decisions without being explicitly programmed for every task.

Simple words: Instead of writing fixed instructions, we give data and machine finds rules.

Traditional programming: Input + Rules → Output

Machine Learning: Input + Output examples → Machine learns Rules

This is the core revolution.

Example: Suppose the data of height and weight
with corresponding "obese/not obese", machine learns relation.

When we give the new person's data (height/weight), then machine predict accordingly.

**4. AI vs ML COMPARISON**

Students often confuse here.

AI: Broad intelligent systems, Human intelligence mimic, Can be rule-based too, Includes robotics, planning, NLP, vision

ML: The subset of AI, which learn from data, pattern/examples, and predict the outcome.

One line: Every ML is AI, but every AI is not ML.

Example: simple chess rule engine can be AI, but if chess system improves from games = ML involved.

**5. WHY MACHINE LEARNING IS NEEDED?**

Question: normal programming se kaam chal jata to ML kyun banaya?

Because some problems have too many hidden patterns.

Example: handwriting recognition, disease prediction, stock trend, fraud detection, recommendation systems

Inme exact human-written rules banana impossible tha.

So we let machine discover hidden mathematical relationships.

That is ML need.

**6. HOW MACHINE LEARNING THINKS (core philosophy)**

Machine learning entirely depends on 3 things.

DATA: past examples

PATTERN: hidden relationship

PREDICTION: future output

Formula: More useful data → Better pattern learning → Better prediction

This is ML heartbeat.

**7. TYPES OF MACHINE LEARNING (overview)**

There are 3 major types.

We will practice deeply later, intro:

**(A) Supervised Learning**

Machine ko input + correct output dono diye jate.

Teacher present hota hai.

Like: Hours Study 2, Marks 40; Hours Study 5, Marks 75

Machine relation learns.

New student study 4 hours → predicted marks can be 60.

This is supervised.

**(B) Unsupervised Learning**

Only data will be given as input, and the machine will find the hidden group by itself.

Example: Divided shopping customers into groups.

No teacher present there, Machine says: these customers are similar or non-similar.

**(C) Reinforcement Learning**

Machine learn through environment via reward or penalty. It's gradually learns best actions then.

Example: game playing bot.
