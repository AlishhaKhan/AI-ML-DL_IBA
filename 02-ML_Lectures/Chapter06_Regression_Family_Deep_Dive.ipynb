**PART A — WHAT IS REGRESSION IN GENERAL?**

Regression means predicting a continuous numerical dependent variable using one or more independent variables.

Simple: some inputs diye jate hain, machine exact number predict karti hai.

Example: marks, salary, house price, temperature

Regression = number prediction.

**PART B — LINEAR REGRESSION**

**1. WHAT IS LINEAR REGRESSION?**

Linear Regression is a supervised learning algorithm used to model the linear relationship between input variables and a continuous output.

Linear relationship means: as X changes, Y changes roughly in a straight-line pattern.

Example: more study hours → more marks.

**2. SIMPLE LINEAR REGRESSION**

When only one input variable predicts one output.

Hours Study → Marks

One X, one Y

**3. LINEAR FUNCTION / EQUATION**

Linear regression tries to fit a best straight line.

y = mx + b

Y = β0 + β1	X

Where:

β₀ = intercept
β₁ = slope
X = input
Y = predicted output

Machine basically best slope and intercept dhoondti.

**4. HOW LINEAR REGRESSION WORKS?**

Suppose data points scattered hain, Model ek straight line draw karta jo points ke sabse close ho, That line should minimize total prediction error. This is called best fit line.

**5. RESIDUAL**

Residual = actual value - predicted value

Residual = y − y^

If actual marks 80 and model predicted 75

Residual = 5

Residual shows prediction kitna off hai.

Small residual = good fit.

**6. ASSUMPTIONS OF LINEAR REGRESSION**

Main assumptions:

relationship linear ho,
residuals random hon,
variance constant ho,
observations independent hon,
no strong multicollinearity (for multiple)

Conceptually:
data line se roughly explain ho sake.

**7. SSE — SUM OF SQUARED ERRORS**

Model har residual square karta and sab add.

SSE=∑(y−y^​)2

Why square? negatives positive ban jaye, large errors punish ho

Goal: SSE minimum

**8. COST FUNCTION**

Cost function = total error measuring function machine minimizes.

For linear regression mostly SSE or mean squared form.

J(β)=1/n∑(y−y^​)2

Lower cost = better line.

**9. GRADIENT DESCENT**

Question: best β values kaise milen?

Machine randomly coefficients choose karti, cost calculate karti, phir downhill move karti to reduce error again and again.

This optimization process = gradient descent.

Imagine hill se neeche utarna to lowest point.

Lowest point = minimum cost.

**PART C — LOGISTIC REGRESSION**

Now important twist: Linear regression numbers predict karti.

But what if output Yes/No ho?, Pass/Fail?, Disease/No disease?

Straight line se unrestricted numbers aayenge, Need probability bounded 0 to1.

That is Logistic Regression.

**1. WHAT IS LOGISTIC REGRESSION?**

Logistic Regression is a supervised learning algorithm used for binary classification by estimating probability.

Output category predict karta, usually 0 or 1.

**2. LINEAR vs LOGISTIC**

Linear: predict any number (-∞ to +∞)

Logistic: predict probability (0 to1)

Then threshold se class.

**3. SIGMOID FUNCTION**

Magic curve that converts any real value into probability.

σ (z) = 1/1 + e−z

Output always between 0 and 1.

If probability >0.5 ⇒ class1
else class0

**4. LOGISTIC FUNCTION PROCESS**

First: z = β₀ + β₁X

Then sigmoid(z)

So line goes into S-shaped probability.

**5. WHY CALLED REGRESSION IF USED FOR CLASSIFICATION?**

Because internally it still estimates coefficients mathematically like regression.

It models log odds using regression equation.

Final use = classification.

Name historical mathematics se aya.

**6. WHY NOT USE MSE HERE?**

Because logistic output nonlinear probability hota, MSE optimization poor behave karti. So logistic usually log loss maximize/minimize karti. Concept enough.

**7. WHEN TO USE LINEAR vs LOGISTIC?**

**Linear Regression:** when target numeric.

**Logistic Regression:** when target binary category.

**PART D — REGRESSION PERFORMANCE METRICS**

Now machine ne predict kar diya. How to check kitna acha? Residual based metrics.

**1. MSE — Mean Squared Error**

Average squared residual, Large errors heavily punish, Lower MSE better.

**2. MAE — Mean Absolute Error**

Absolute errors average, Easy interpret, Average kitna miss kiya.

**3. RMSE**

Square root of MSE, Same units as original target.

**4. MAPE**

Percentage error, Average percentage miss.

**5. MPE**

Signed percentage error, Shows under/over prediction bias.

**6. R² SCORE**

How much variance model explains, Range near 1 = good, If R²=.90 means model explains 90% variability.

**7. ADJUSTED R²**

R² ka improved version for multiple inputs, Penalizes useless extra variables, Used when many features.
