```python
# remove unnecessary column
df = df.drop(columns='Unnamed: 0')
df
```

**Data Cleaning**

```python
df.isnull().sum()
```

```python
# 1. Listwise Deletion
df_clean = df[df.isnull().sum(axis=1) <4].reset_index(drop=True).copy()
```

```python
df_clean.head()
```

```python
HmMode = df_clean['Humidity'].mode()[0]
HdMode = df_clean['Holiday'].mode()[0]

df_clean['Humidity'] = df_clean['Humidity'].fillna(HmMode)
df_clean['Holiday'] = df_clean['Holiday'].fillna(HdMode)

df_clean.head()
```

```python
df_clean.columns.tolist()
```

This code fills missing values (NaNs) in specific columns to prevent gaps in the dataset and ensure cleaner, more complete data for analysis and model training

```python
# 3. Linear Interpolation
df_clean['Temp'] = df_clean['Temp'].interpolate(method='linear')
df_clean.head()
```

```python
# 4. Forward/Backward Fill
df_clean['Outlook'] = df_clean['Outlook'].ffill()
df_clean.head()
```

.fillna(-1): It looks for every missing value (NaN) in the 'Wind' column and replaces it with the number -1.

```python
df_clean['Wind'] = df_clean['Wind'].fillna(-1)
```
