```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'/content/RegData.csv')
df.sample(5)
```

```python
col2 = df['GPA']
```

```python
import warnings
warnings.filterwarnings('ignore')
def dist2Chrt(c2):
  plt.figure(figsize=(16,5))
  plt.subplot(1,2,1)
  sns.distplot(c2)
  plt.show()
```

```python
dist2Chrt(col2)
```

```python
def dist2Chrt(c2):
  plt.figure(figsize=(16,5))
  plt.subplot(1,2,2)
  sns.distplot(c2)
  plt.show()
```

```python
dist2Chrt(col2)
```

```python
sns.boxplot(col2)
```

```python
q1 = col2.quantile(0.25)
q2 = col2.quantile(0.50)
q3 = col2.quantile(0.75)
print("Min = 0.0")
print("percentile 25% =" , q1)
print("percentile 50% =" , q2)
print("percentile 70% =" , q3)
```

```python
iqr = q3 - q1
iqr
```

```python
iqr = q3 - q1
float(iqr)
```

```python
iqrVal = 1.5 * iqr
lower_limit = q1 - iqrVal
upper_limit = q3 + iqrVal
print("Lower Limit:",lower_limit)
print("Upper Limit:",upper_limit)
```

```python
df[col2 < lower_limit].count()
```

```python
df[col2 > upper_limit].count()
```

```python
df[col2 > upper_limit]
```

```python
new_df = df[(col2 > lower_limit) & (col2 < upper_limit)]
new_df.shape
```

```python
sns.boxplot(new_df['GPA'])
```

```python
new_df.to_csv('/content/RegData_new.csv')  ## here we create new file by code
```

```python
plt.figure(figsize=(16,8))

# Original Data
plt.subplot(2,2,1)
sns.distplot(df['GPA'])

plt.subplot(2,2,2)
sns.boxplot(df['GPA'])

# Cleaned Data
plt.subplot(2,2,3)
sns.distplot(new_df['GPA'])

plt.subplot(2,2,4)
sns.boxplot(new_df['GPA'])

plt.show()
```
