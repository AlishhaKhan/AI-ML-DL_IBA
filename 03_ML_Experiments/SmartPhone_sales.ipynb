```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df01 = df.iloc[:, 1:]
```

```python
df01
```

```python
cols = df01.columns
df01
```

```python
df02 = df01.sum()
df02
```

**Bar Plot**

```python
df02.values
```

```python
plt.figure(figsize = (15, 6))
plt.xticks(range(len(df02)), cols)
plt.title("Bar Plot")
plt.xlabel("Phone")
plt.ylabel("Sale")
# chrt = plt.bar(range(len(df02)), df02, color='royalblue', alpha=0.7)
chrt = sns.barplot(x=df02.index, y=df02.values)
for p in chrt.patches:
  chrt.annotate(format(p.get_height(), ',.0f'),
                 (p.get_x() + p.get_width() / 2., p.get_height()),
                 ha = 'center', va = 'center', xytext = (0, 9),
                 textcoords = 'offset points')
```

**Line graph**

```python
plt.figure(figsize=(10, 4))

plt.xticks(rotation=45, ha="right")
plt.title("Bar Plot")
plt.xlabel("Phone")
plt.ylabel("Sale")

# Convert index to numeric positions
x = range(len(df02))

# Plot
plt.plot(x, df02.values, marker='o', color='pink')

# Labels on points
for i, (a, b) in enumerate(zip(df02.index, df02.values)):
    plt.text(i, b, str(b), ha='center')

# Set actual labels
plt.xticks(x, df02.index)

plt.tight_layout()
plt.show()
```

**Pie Chart**

```python
plt.figure(figsize = (10, 10))
plt.title('pie chart')
plt.pie(df02, labels=cols, labeldistance=1.15, autopct='%1.2f%%', wedgeprops={'linewidth':3, 'edgecolor':'white'})
```

**Distribution Plot**

```python
chrt = sns.displot(x=df02.index, y=df02.values)
```

```python
ax = df.plot.bar(stacked=True, figsize=(15, 7), rot=0)

plt.title("Stacked Chart")
plt.xlabel("Quarter")
plt.ylabel("Sale")
plt.xticks(range(len(df)), df.Quarter)

for c in ax.containers:
    ax.bar_label(c, label_type='center')

ax.legend(title='Categories', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```

**UnStacked Bar Chart**

```python
chrt = df01.plot(kind='bar', figsize=(15, 7))

plt.xticks(range(len(df01)), df.Quarter)

plt.title("UnStacked Chart")
plt.xlabel("Quarter")
plt.ylabel("Sale")

for p in chrt.patches:
    chrt.annotate(format(p.get_height(), '.0f'),
                  (p.get_x() + p.get_width() / 2., p.get_height()),
                  ha='center', va='center',
                  xytext=(0, 5),
                  textcoords='offset points')

plt.show()
```
