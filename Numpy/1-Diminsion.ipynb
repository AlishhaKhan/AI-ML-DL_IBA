*1D Numpy in Python*

Create a Python List as follows:

```python
a = [0, 1, 2, 3, 4]
```

We can access the data via an index:

```python
print("a[0]:", a[0])
print("a[1]:", a[1])
print("a[2]:", a[2])
print("a[3]:", a[3])
print("a[4]:", a[4])
```

A numpy array is similar to a list. It's usually fixed in size and each element is of the same
type. We can cast a list to a numpy array by first importing numpy:

```python
import numpy as np
```

```python
# we can cast the list
a = np.array([0, 1, 2, 3, 4])
a
```

As with lists, we can access each element via a square bracket:

```python
print("a[0]:", a[0])
print("a[1]:", a[1])
print("a[2]:", a[2])
print("a[3]:", a[3])
print("a[4]:", a[4])
```

*Assign Value*

We can change the value of the array, consider the array c :

```python
c = np.array([20, 1, 2, 3, 4])
c
```

We can change the first element of the array to 100 as follows:

```python
c[0] = 100
c
```

We can change the 5th element of the array to 0 as follows:

```python
c[4] = 400
c
```

*Slicing*

Like lists, we can slice the numpy array, and we can select the elements from 1 to 3 and
assign it to a new numpy array d as follows:

```python
d = c[1:4]
d
```

We can assign the corresponding indexes to new values as follows:

```python
# Set the third element and fourth element to 300 and 400
c[3:5] = 300, 400
c
```

*Assign Value with List*

Similarly, we can use a list to select a specific index. The list ' select ' contains several
values:

```python
select = [0, 2, 3]
```

We can use the list as an argument in the brackets. The output is the elements
corresponding to the particular index:

0 -------- 2 -- 3

[100, 1, 2, 300, 400]

```python
d = c[select]
d
```

We can assign the specified elements to a new value. For example, we can assign the

```python
c[select] = 100000
c
```

*Linspace*

A useful function for plotting mathematical functions is "linespace". Linespace returns
evenly spaced numbers over a specified interval. We specify the starting point of the
sequence and the ending point of the sequence. The parameter "num" indicates the
Number of samples to generate, in this case 5:

```python
# Makeup a numpy array within [-2, 2] and 5 elements
np.linspace(-2, 2, num=5) # start, stop, number of values
```

If we change the parameter num to 9, we get 9 evenly spaced numbers over the interval
from -2 to 2:

```python
np.linspace(-2, 2, num=9)
```

We can use the function line space to generate 100 evenly spaced samples from the
interval 0 to 2π:

```python
# Makeup a numpy array within [0, 2π] and 5 elements
x = np.linspace(0, 2*np.pi, num=5)
x
```

We can apply the sine function to each element in the array x and assign it to the array
y :

```python
y = np.sin(x)
y
```
