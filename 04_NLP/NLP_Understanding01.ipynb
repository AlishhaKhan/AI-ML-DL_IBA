**🔹 PRACTICAL 2 — TOKENIZATION CODE**

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
text = 'Natural Language Preprocessing is interesting.'

words = nltk.word_tokenize(text)          # sentence broken into words
print(words)
```

**🔹 PRACTICAL 3 — STEMMING CODE**

```python
from nltk.stem import PorterStemmer

ps = PorterStemmer()          # words convert into rough root

print(ps.stem("playing"))
print(ps.stem("running"))
print(ps.stem("happiness"))
```

**🔹 PRACTICAL 4 — LEMMATIZATION CODE**

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
```

```python
lm = WordNetLemmatizer()
                                          # Proper dictionary root form.
print(lm.lemmatize("running", pos='v'))
print(lm.lemmatize("better", pos='a'))
```

**🔹 PRACTICAL 5 — BAG OF WORDS (sklearn)**

```python
from sklearn.feature_extraction.text import CountVectorizer

docs = [
    "I love AI",
    "I love ML",
    "I love AI and ML"
]                               # Text converted to count vectors

cv = CountVectorizer()
X = cv.fit_transform(docs)

print(cv.get_feature_names_out())
print(X.toarray())
```

**🔹 PRACTICAL 6 — TF-IDF**

```python
from sklearn.feature_extraction.text import TfidfVectorizer

docs = [
    "I love AI",
    "I love ML",
    "I love AI and ML"
]                                # Text converted to weighted importance vectors

tfidf = TfidfVectorizer()
X = tfidf.fit_transform(docs)

print(tfidf.get_feature_names_out())
print(X.toarray())
```

**🔹 PRACTICAL 7 — POS TAGGING**

```python
nltk.download('averaged_perceptron_tagger_eng')
```

```python
sentence = "The big brown dog barked loudly."
words = nltk.word_tokenize(sentence)
pos_tags = nltk.pos_tag(words)
                                      # Every word got grammar label
print(pos_tags)
```

**🔹 PRACTICAL 8 — CHUNKING / REGEX PARSER**

```python
reg_exp = "NP: {<DT>?<JJ>*<NN>}"

rp = nltk.RegexpParser(reg_exp)
result = rp.parse(pos_tags)        # Machine groups noun phrase

print(result)
```

**🔹 PRACTICAL 9 — NAMED ENTITY RECOGNITION**

```python
nltk.download('maxent_ne_chunker_tab')
nltk.download('words')
```

```python
chunked = nltk.ne_chunk(pos_tags)      # Machine tries to detect names, locations, organizations
print(chunked)
```
