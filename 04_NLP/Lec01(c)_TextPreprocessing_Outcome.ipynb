```python
import textpreprocess as pp
```

```python
# 1) Change Case
text = "Hello World!"
print(pp.ChangeCase(text))
```

```python
# 2) Remove punctuation and special characters
text = " Hello World! This is?* ❤️ example. "
print(pp.RmovePunctuationSpChar(text))
```

```python
# 3. Remove URLs
text = "Visit our website at https://www.example.com"
print(pp.RemoveURL(text))
```

```python
# 4. Remove HTML tags
text = "<p>This is a <strong>sample</strong> sentence.</p>"
print(pp.RemoveHTML(text))
```
