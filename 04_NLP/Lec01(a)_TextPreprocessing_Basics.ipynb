```python
text = "Hello, world! This is?* 💜an&/|~^+%\" example-of text preprocessing."

punctuation_pattern = r'[^\w\s]'

text_cleaned = re.sub(punctuation_pattern, '', text)
                                                           # function, special character remove kar de ge
print(text_cleaned)
```

```python
from nltk.corpus import stopwords
```

```python
# remove english stopwords functions
def remove_stopwords(text, language):
  stop_words = set(stopwords.words(language))
  text = text.lower()                           # convert karna hai function mein
  word_tokens = text.split()                                # split ko comment karna hai assignment k time
  filtered_text = [word for word in word_tokens if word not in stop_words]     # list comprehension
  print(language)
  print(filtered_text)
```

```python
en_text = "This is a sample sentence and we are goingto remove the stopwords from this"
remove_stopwords(en_text, "english")
```

**URLS**

```python
def remove_urls(text):
  url_pattern = re.compile(r'https?://\S+|www.\. \S+')       # already bana huwa hai
  return url_pattern.sub(r'',text)
```

```python
text = "I hope it wil be a useful article for you. Follow me: https://abc.com/@def"
remove_urls(text)
```

**HTML tags**

```python
text = """<html><div>
<h1>Natural Language Processing</h1>
<p>Text Preprocessing for NLP</p>
<a href="https://abc.com/@def">ABC account</a>
</div></html>"""

html_tags_pattern = r'<.*?>'
                                                                         # html tags ka fuction bane ga
text_without_html_tags = re.sub(html_tags_pattern, '', text)

print(text_without_html_tags)
```
