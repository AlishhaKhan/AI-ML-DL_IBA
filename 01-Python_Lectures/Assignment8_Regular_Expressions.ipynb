**15. Regular Expression**

**Definition**

Regular Expression (RegEx) is a sequence of characters used to search, match and manipulate patterns in text.

**Purpose**

* Find specific pattern in text

* For validation (email, phone number, password)

* Search and replace operations

**Common Pattern**

* \d - digit(0-9)

* \w - word(a-z)

* , - any character

* ^ - start of string

* $ - end of string

* * - 0 or more times

* + - 1 or more times

```python
import re
```

```python
text = "My number is 12345"
# find digit
result = re.findall(r"\d+", text)
print(result)
```

```python
text = "hello123"
if re.search(r"\d", text):
  print("Number found")
```

```python
stmt = 'pythonnn example:regx123 egg' # stmt : jis text pe hum pattern apply kar rahe hain
```

```python
ptrn = r'example:\w\w\w\w'   # fixed pattern match
output = re.search(ptrn, stmt)
if output:
 print ('pattern found:', output.group())
else:
 print ('not found')
```

```python
ptrn = 'e:r'      # direct string match, exact string find kare ga
output = re.search(ptrn, stmt)
if output:
 print ('pattern found:', output.group())
else:
  print("not found")
```

```python
ptrn = '..:..'   # 2 characters + : + 2 characters
output = re.search(ptrn, stmt)
if output:
 print ('pattern found:', output.group())
else:
  print('not found')
```

```python
ptrn = r'\d\d\d'  # 3 continuous digit
output = re.search(ptrn, stmt)
if output:
 print ('pattern found:', output.group())
else:
  print('not found')
```

```python
ptrn = 'on+' # n+ = 1 ya zyada n
output = re.search(ptrn, stmt)
if output:
 print ('pattern found:', output.group())
else:
  print('not found')
```

```python
ptrn = r'^p\w+'   # ^ = start of string, \w+ = word characters (1 ya zyada)
output = re.search(ptrn, stmt)
if output:
 print ('pattern found:', output.group())
else:
  print('not found')
```

```python
ptrn = r'e\w+' # e se start hone wala word
output = re.search(ptrn, stmt)
if output:
 print ('pattern found:', output.group())
else:
  print('not found')
```

```python
stmt2 = 'my email address is alisha.khan@gmail.com'
```

```python
ptrn = r'\w+\.\w+@\w+\.\w+'
output = re.search(ptrn, stmt2)
if output:
 print ('pattern found:', output.group())
else:
 print ('not found')
```

```python
stmt2 = 'my email address is alisha_khan@gmail.com'
```

```python
ptrn = r'\w+\.\w+@\w+\.\w+'
output = re.search(ptrn, stmt2)
if output:
 print ('pattern found:', output.group())
else:
 print ('not found')
```

```python
ptrn = r'[\w+]+@[\w.]+'   # [] = set, \w+ = letter\number\underscore, . = allow kia
output = re.search(ptrn, stmt2)
if output:
 print ('pattern found:', output.group())
else:
 print ('not found')
```

```python
stmt2 = 'my email address is alisha.khan@gmail.com'
```

```python
ptrn = r'[\w+]+@[\w.]+'
output = re.search(ptrn, stmt2)
if output:
 print ('pattern found:', output.group())
else:
 print ('not found')

 # kyun? pattern ne sirf valid part uthaya, pura email nahi
 # exam point : RegEx partial match bhi return kar sakta hai
```

```python
ptrn = r"e\w+" # saare matches return karega
output = re.findall(ptrn,stmt)
print(output)
```

```python
ptrn = r"\s" # \s = space
output = re.split(ptrn,stmt)
print(output)
```

```python
ptrn = r"p" # jahan jahan p milega, split karega
output = re.split(ptrn,stmt)
print(output)
```

```python
ptrn = r"py\w+"  # multiple match loop
output = re.findall(ptrn,stmt)
for p in output:
 print(p)
```

```python
stmt = "pythonnn example:regx123 python"
```

```python
ptrn = r"py\w+" # final
output = re.findall(ptrn,stmt)
res = []
for p in output:
 res.append(p)
print(res)
```
