*Car System*

Vehicle (base class)

Car (subclass)

Engine, Battery, AC (component classes)

           +------------------+
           |     Vehicle      |
           +------------------+
           | - vehType        |
           | - vStatus        |
           | - vehID          |
           | - wheels         |
           | - make           |
           | - model          |
           | - color          |
           | - engine         |
           | - battery        |
           | - ac             |
           +------------------+
           | + move()         |
           | + get_info()     |
           | + setters/getters|
           +------------------+
                    ▲
                    |
               Inheritance
                    |
           +------------------+
           |       Car        |
           +------------------+
           |  + move()        |
           +------------------+

       Aggregation (Weak Relationship):

       Vehicle HAS-A Engine, Battery, AC

           +---------+     +----------+     +------+
           | Engine  |     | Battery  |     |  AC  |
           +---------+     +----------+     +------+
           | engnID  |     | btryID   |     | acID |
           | status  |     | status   |     |status|
           +---------+     +----------+     +------+

```python
# Engine Class
class Engine:
    def __init__(self, engnID):
        self.__engnID = engnID
        self.__estatus = "Working"

    def set_estatus(self, status):
        self.__estatus = status

    def get_estatus(self):
        return self.__estatus

    def get_engnID(self):
        return self.__engnID

 #  Battery Class
class Battery:
    def __init__(self, btryID):
        self.__btryID = btryID
        self.__btrystatus = "Working"

    def set_btrystatus(self, status):
        self.__btrystatus = status

    def get_btrystatus(self):
        return self.__btrystatus

    def get_btryID(self):
        return self.__btryID

  #  AC Class
class AC:
    def __init__(self, acID):
        self.__acID = acID
        self.__estatus = "Working"

    def set_acstatus(self, status):
        self.__estatus = status

    def get_acstatus(self):
        return self.__estatus

    def get_acID(self):
        return self.__acID
```

```python
#  Vehicle (Base Class)
class Vehicle:
    def __init__(self, vehID, wheels, make, model, color, engine, battery, ac=None):
        self.__vehType = "Vehicle"
        self.__vStatus = "Working"

        self.__vehID = vehID
        self.__wheels = wheels
        self.__make = make
        self.__model = model
        self.__color = color

        self.__engine = engine
        self.__battery = battery
        self.__ac = ac

    def set_vehType(self, vehType):
        self.__vehType = vehType

    def set_engn_status(self, sts):
        self.__engine.set_estatus(sts)
        self.__vStatus = sts

    def set_btry_status(self, sts):
        self.__battery.set_btrystatus(sts)
        self.__vStatus = sts

    def set_ac_status(self, sts):
        if self.__ac:
            self.__ac.set_acstatus(sts)

    def get_vehType(self):
        return self.__vehType

    def get_vehID(self):
        return self.__vehID

    def get_vStatus(self):
        return self.__vStatus

    def get_engn_status(self):
        return self.__engine.get_estatus()

    def get_btry_status(self):
        return self.__battery.get_btrystatus()

    def get_ac_status(self):
        return self.__ac.get_acstatus()

    def get_info(self):
      return f"{self.__color} {self.__wheels}-wheeler {self.__make} {self.__model}"

    def move(self, txt="The"):
       if self.__vStatus != "Working":
            raise Exception(f"The {self.__vehType} cannot move. Status: {self.__vStatus}")
            print(f"{txt} {self.__vehType} is moving.")
```

```python
#  Car (Subclass)
class Car(Vehicle):
    def __init__(self, vehID, wheels, make, model, color, engine, battery, ac=None):
        super().__init__(vehID, wheels, make, model, color, engine, battery, ac)
        self.set_vehType("Car")

    def move(self, txt="The"):
        if self.get_vStatus() != "Working":
            raise Exception(f"The {self.get_vehType()} cannot move. Engine/Battery issue.")
        print(f"{txt} {self.get_vehType()} is being driven.")
```

```python
#  Objects
eng = Engine(1)
bat = Battery(1)
ac = AC(1)

my_car = Car(101, 4, "Toyota", "Corolla", "Black", eng, bat, ac)
```

```python
# Test
print(my_car.get_info())
my_car.move("My")

# Create problem
my_car.set_engn_status("Problem")

try:
    my_car.move("My")
except Exception as e:
    print(e)

# Fix again
my_car.set_engn_status("Working")
my_car.move("My")
```
