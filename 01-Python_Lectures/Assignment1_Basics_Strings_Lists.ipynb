1) Print student name, course, and year in a single line. Explanation: Display multiple pieces of information in one line using print. Example Values: Name = Ali, Course = Python, Year = 2026 Expected Output: Name = Ali | Course = Python | Year = 2026

```python
name = "Ali"
course = "Python"
year = 2026

print("Name =", name, "| Course =", course, "| Year =", year)
```

2) Print the same output using the f-string. Explanation: Use f-strings to insert variable values inside a string. Example Values: Name = Sara, Course = Data Science, Year = 2025 Expected Output: Name = Sara | Course = Data Science | Year = 2025

```python
name = "Sara"
course = "Data Science"
year = 2025

print(f"Name = {name} | Course = {course} | Year = {year}")
```

3) Demonstrate dynamic typing using a single variable. Explanation: Show that a variable can store different data types. Example Values: 10 -> 10.5 -> Python

```python
x = 10
print(x, type(x))

x = 10.5
print(x, type(x))

x = "Python"
print(x, type(x))
```

4) Take two numbers and print sum, floor division, and reminder. Explanation: Perform basic arithmetic operations. Example Values: Number1 = 15, Number2 = 4 Expected Results: Sum = 19, Floor Division = 3, Reminder = 3

```python
num1 = 15
num2 = 4

print("Sum =", num1 + num2)
print("Floor Division =", num1 // num2)
print("Remainder =", num1 % num2)
```

5) Take birth year and calculate age. Explanation: Calculate age using the current year. Example Values: Birth Year = 2000, Current Year = 2026 Expected Age = 26

```python
birth_year = 2000
current_year = 2026

age = current_year - birth_year
print("Age =", age)
```

6) Swap two variables without using a temporary variable. Explanation: Exchange values using Python's swapping technique. Example Values: a = 5, b = 10 After Swap: a = 10, b = 5

```python
a = 5
b = 10

a, b = b, a

print("After Swap:")
print("a =", a)
print("b =", b)
```

7) Swap two variables using a temporary variable. Explanation: Use a third variable to swap values. Example Values: x = 3, y = 7 After Swap: x =7, y =3

```python
x = 3
y = 7

temp = x
x = y
y = temp

print("After Swap:")
print("x =", x)
print("y =", y)
```

8) Demonstrate augmented assignment operators. Explanation: Modify a variable using operators like += and *=. Example Values: Start = 5 -> After += 10 -> After *= 2

```python
value = 5
print("Start =", value)

value += 10
print("After += 10 =", value)

value *= 2
print("After *= 2 =", value)
```

21) Add and insert elements in a list. Explanation: Add items at the end or at a specific position. Example: ['Python', 'Java'] -> ['Python', 'Java', 'C++']

```python
languages = ['Python', 'Java']

# Add at end
languages.append('C++')

# Insert at specific position
languages.insert(1, 'Go')

print(languages)
```

22) Remove elements using remove and pop. Explanation: Delete items from a list. Example: ['Python', 'Java', 'C++'] -> ['Python', 'C++']

```python
languages = ['Python', 'Java', 'C++']

# Remove by value
languages.remove('Java')

# Remove last element
languages.pop()

print(languages)
```

Example:
List = [12, 4, 19, 7]
