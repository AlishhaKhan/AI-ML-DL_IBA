*Join (merge)*

```python
e = emp.copy() # pd.read_csv(r"D://DataSets//emp.csv")
d = pd.read_csv("dept.csv")
print(e.columns)
print(d.columns)
```

```python
e.head()
```

```python
d
```

```python
# JOIN all same column(s) name
```

```python
pd.merge(e, d)
```

```python
# JOIN specific same column(s) name
# Joining Type
```

*inner join*

Same column Name in both dataframes

```python
pd.merge(e, d, on='deptno', how='inner')
```

Different Column Names in both dataframes

```python
pd.merge(e, d, left_on='deptno', right_on='deptno', how='inner')
```

*Outer Join*

```python
pd.merge(e, d, how='right') # 'left'; 'outer'
```

```python
pd.merge(e, d, on='deptno', how='outer', indicator = True)
```

```python
pd.merge(e, d, on='deptno', how='inner', validate = 'many_to_one') # 'on_to_many'
```

*Self Join: Two columns of One dataframe*

```python
e = emp
m = emp
```

```python
pd.merge(e, m, left_on='mgr', right_on='empno', how='inner')[['empno_x', 'ename_x', 'empno_y', 'ename_y']]
# SELECT e.empno, e.ename, e.mgr, m.ename
# FROM emp e INNER JOIN emp m ON e.mgr= m.empno
```

```python
pd.merge(e, m, left_on='mgr', right_on='empno', how='outer')[['empno_x', 'ename_x', 'empno_y', 'ename_y']]
```

```python
pd.merge(e, m, left_on='mgr', right_on='empno', how='left')[['empno_x', 'ename_x', 'empno_y', 'ename_y']]
```

```python
pd.merge(e, m, left_on='mgr', right_on='empno', how='right')[['empno_x', 'ename_x', 'empno_y', 'ename_y']]
```

*Set Operations*

```python
j20 = emp.job[(emp.deptno==20)].unique()
print(j20)
j30 = emp.job[(emp.deptno==30)].unique()
print(j30)
```

```python
j20 = set(j20)
j20
```

```python
j30 = set(j30)
j30
```

*union*

```python
union_set = j20.union(j30)
union_set
```

```python
union_df = pd.DataFrame(list(union_set), columns=['job'])
union_df
```

*intersection*

```python
intersect_set = j20.intersection(j30)
intersect_set
```

*minus*

```python
minus_set2030 = j20.difference(j30)
minus_set2030
```

```python
minus_set3020 = j30.difference(j20)
minus_set3020
```

*symmetric difference*

```python
symdiff_set = j20.symmetric_difference(j30)
symdiff_set
```
