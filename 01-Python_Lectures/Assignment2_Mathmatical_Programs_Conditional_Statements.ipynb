6) Find the area of triangle a = 5, b = 6, c = 7, also ask the input from the user.
hf.PNG

```python
a = float(input("Enter side a: "))
b = float(input("Enter side b: "))
c = float(input("Enter side c: "))

s = (a + b + c) / 2

area = (s * (s-a) * (s-b) * (s-c)) ** 0.5

print("Area of triangle =", area)
```

7) Solve the quadratic equation ax**2 + bx + c = 0


qf.PNG

```python
a = float(input("Enter a: "))
b = float(input("Enter b: "))
c = float(input("Enter c: "))

d = b**2 - 4*a*c

if d > 0:
    root1 = (-b + (d ** 0.5)) / (2*a)
    root2 = (-b - (d ** 0.5)) / (2*a)
    print("Two Real Roots:", root1, root2)

elif d == 0:
    root = -b / (2*a)
    print("One Real Root:", root)

else:
    print("Roots are complex")
```

11) Demonstrate Nested if Statement.
Explanation:
Write a program to demonstrate nested if statements.

```python
i = 10

if i == 10:
    if i < 15:
        print("i is smaller than 15")

    if i < 12:
        print("i is smaller than 12 too")
else:
    print("i is greater than 15")
```
