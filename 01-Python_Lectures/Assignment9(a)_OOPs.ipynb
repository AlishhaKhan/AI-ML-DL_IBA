**Step 1 : Creating a Class**

*Class:* A general template. A blueprint (like a plan for making objects).

 Class is like a common noun.

 *Constructor:* _init_ is a constructor, it run when we create an object.

 A constructor is a special method that run automatically when we create a new object from a class.

 It helps to setting up the initial values of object(like name or age of a person).

 In Python, the constructor method is named _init_, which stands of initialize.

 *Attributes:* (Properties, Features)

 self.wheel = wheel assign the values to the object.

 Attributes are like the adjective.

 *Actions:* (behaviour) of an objects.

 *Methods:* they are like verbs, they do something.

```python
class car:
  # This is an attribute ( class variable) common to all cars.
  vehType = "Car"
  # This is a special method called constructor.
  # It run when we create a new car object.
  def __init__(self, wheels, make, model, color):
    self.wheels = wheels  # instance attribute
    self.make = make      # instance attribute
    self.model = model    # instance attribute
    self.color = color    # instance attribute

  # This is a method ( an action the car can do)
  def move(self, txt="The"):
    print(f"{txt} {self.vehType} is being driven.")
```

**Object Oriented Programming (OOPs)**

The way of writing software that revolve around an object.

These objects can store data and their actions (methods).

Rather than concentrating solely on processes and logic, OOP encourage the structure of code around these objects.

This approach makes it easier to create modular, reusable and scalable software design.

**The four main principles of OOPs.**

* Encapsulation

* Abstraction

* Inheritance

* Polymorphism

**Encapsulation:**

Encapsulation helps to protect the data inside an object.

Think of it like keeping certain details private, allowing only controlled access to them.

That is, instead of directly changing or viewing the data, you interact with through specific method.

This ensures that the data is safe from unintended changes.

```python
class Car:
  __vehType = "Car"     # private class attribute
  def __init__(self, wheels, make, model, color):
    self.__wheels = wheels   # private instance attribute
    self.__make = make      # private instance attribute
    self.__model = model     # private instance attribute
    self.color = color     # public instance attribute

  def move(self, txt="The"):
    print(f"{txt} {self.__vehType} is being driven.")
```

**Abstraction**

Abstraction allows you to show only the important details of an object or system, while hiding the complex parts that the user doesn't need to see.

Think of it like driving a car: you only need to know how to use the steering wheel, the gearshift, the gas pedal and the break.

You don't need to understand how the engine needs to work internally.

In programming, abstraction helps you focus on what something does rather than how it works internally.

**Encapsulation: Focuses on building data and restricting access.**

**Abstraction: Focuses on hiding complexity and exposing only necessary details.**

*Add get_info() method*

```python
class Car:
    __vehType = "Car"  # Private class attribute

    def __init__(self, wheels, make, model, color):
        self.__wheels = wheels
        self.__make = make
        self.__model = model
        self.color = color

    def get_vehType(self):
        return Car.__vehType   # IMPORTANT (class variable access)

    def get_wheels(self):
        return self.__wheels

    def get_info(self):
        info = f"{self.color} {self.__wheels}-wheeler, {self.__make} {self.__model}"
        return info

    def move(self, txt="The"):
        print(f"{txt} {Car.__vehType} is being driven.")  #  IMPORTANT


# Objects
my_car = Car(4, "Toyota", "Corolla", "Black")

# Testing
print(my_car.get_vehType())
print(my_car.get_wheels())
print(my_car.get_info())
my_car.move("My")
```

**Inheritance**

Inheritance, let you create a new class by using an existing class.

The new class(called the child/derived class)gets all the attributes and methods from the existing class(called the parent/base class).

This allows you to reuse code and build upon what you've already written without starting from scratch.

*Base Class/Parent Class*

```python
class Vehicle:

# This is a special method called a Constructor.
# It runs when you create a new object.
 def __init__(self, wheels, make, model, color):
  self.__vehType = "Vehicle" # __ Private instance attribute
  self.__wheels = wheels # __ Private instance attribute
  self._make = make # _ Protected instance attribute
  self._model = model # _ Protected instance attribute
  self.color = color # Public instance attribute

 def set_vehType(self, vehType):
  self.__vehType = vehType # A public method to set the private attribute

 def get_vehType(self):
  return self.__vehType # A public method to access the private attribut

 def get_wheels(self):
  return self.__wheels # A public method to access the private attribute

 def get_info(self):
  infor = f"{self.color} {self.__wheels}-wheeler, {self._make} {self._model}"
  return infor

 def move(self, txt="The"):
   print(f"{txt} {self.get_vehType()} is being driven.")
```

*Derived Class/Child Class*

```python
class Car(Vehicle):
  def __init__(self, wheels, make, model, color):
    super().__init__(wheels, make, model, color)
    self.set_vehType("Car")

class MotorBike(Vehicle):
  def __init__(self, wheels, make, model, color):
   super().__init__(wheels, make, model, color)
   self.set_vehType("Motor Bike")
```

**Duck Typing**

Duck typing is a concept where the type of an object is determined by its behavior (methods and properties) rather than its class.

*"If it look like a duck, swims like a duck, and quack like a duck, than it probably is like a duck"*

In short; If it behaves like a duck, it is a duck.

**Name Mangling**

Name mangling is a mechanism in Python where variables with double underscores (__var) are internally renamed to avoid accidental access or modification.

```python
class Vehicle:

# This is a special method called a Constructor.
# It runs when you create a new object.
  def __init__(self, wheels, make, model, color):
    self.__vehType = "Vehicle" # __ Private instance attribute
    self.__wheels = wheels # __ Private instance attribute
    self._make = make # _ Protected instance attribute
    self._model = model # _ Protected instance attribute
    self.color = color # Public instance attribute

  def get_wheels(self):
   return self.__wheels # A public method to access the private attribute

  def get_info(self):
    infor = f"{self.color} {self.__wheels}-wheeler, {self._make} {self._model}"
    return infor

  def move(self, txt="The"):
    print(f"{txt} {self.get_vehType()} is moving.")
```

**Concrete Class**

A concrete class is a class that provides complete implementation of all its methods and can be instantiated (object can be created from it).

```python
class Vehicle:
  # This is a special method called a constructor.
  # It run when you create a new object.

  def __init__(self, wheels, make, model, color):
    self.__vehType = "Vehicle" # __ Private instance
    self.__wheels = wheels # __ Private instance
    self._make = make # _ Protected instance
    self._model = model # _ Protected instance
    self.color = color # Public instance

  def set_vehType(self, vehType):
    self.__vehType = vehType  # A public method to set the private attribute

  def get_vehType(self):
    return self.__vehType      # A public method to access private attribute

  def get_wheels(self):
    return self.__wheels       # A public method to access private attribute

  def get_info(self):
    infor = f"{self.color} {self.__wheels} wheeler {self._make} {self._model}"
    return infor

  def move(self, txt="The"):
    print(f"{txt} {self.get_vehType()} is moving.")
```

Creating an Object of the Concrete Class

```python
veh = Vehicle(4, "Toyota", "Corolla", "Black")
```

```python
info = veh.get_info()
print(info)
```

```python
veh.move("My " + info)
```

**Abstract Class and Abstract Method**

*Abstract Class:* An abstract class is a class that cannot be instantiated and is used as a blueprint for other classes. It may contain abstract methods.

*Abstract Method:* An abstract method is a method declared in an abstract class without implementation, which must be implemented by its child classes.

📘 Abstract Class vs Abstract Method

👉 Abstract Class:

A class that contains at least one abstract method

Cannot be instantiated

👉 Abstract Method:

A method without body (no implementation)

Must be overridden in child class

```python
from abc import ABC, abstractmethod

class Vehicle(ABC):
  # This is a special method called a Constructor.
  # It runs when you create a new object.
  def __init__(self, wheels, make, model, color):
   self.__vehType = "Vehicle" # __ Private instance attribute
   self.__wheels = wheels # __ Private instance attribute
   self._make = make # _ Protected instance attribute
   self._model = model # _ Protected instance attribute
   self.color = color # Public instance attribute

  def set_vehType(self, vehType):
   self.__vehType = vehType # A public method to set the private attribute

  def get_vehType(self):
   return self.__vehType # A public method to access the private attribut

  def get_wheels(self):
   return self.__wheels # A public method to access the private attribute

  def get_info(self):
    infor = f"{self.color} {self.__wheels}-wheeler, {self._make} {self._model}"
    return infor

@abstractmethod
def move(self, txt="The"):
  pass
```
