**1. Strong Relationship (Composition)**

A strong relationship (composition) is a relationship where one object completely owns another object, and the dependent object cannot exist independently.

🧠 Key Terms

Container/Base Class → main class

Component Class → dependent class

Sub Class → component inside container

Example:

👉 Car aur Engine

Car = container

Engine = component

👉 Agar car destroy ho jaye → engine ka koi independent use nahi

```python
# Example:
class Engine:
    def start(self):
        print("Engine started")

class Car:
    def __init__(self):
        self.engine = Engine()  # strong relation

    def start_car(self):
        self.engine.start()
```

📌 Set to Problem → Set back to Working

👉 Problem:

Car ke bina engine ka koi use nahi

👉 Working:

Engine automatically car ke saath create hota hai

*Component Class*

```python
class Engine:
 def __init__(self, engnID):
  self.__engnID = engnID
  self.__estatus = "Working"
 def set_estatus(self, status):
  self.__estatus = status
 def get_engnID(self):
   return self.__engnID
 def get_estatus(self):
   return self.__estatus
```

*Container/Base Class*

```python
class Vehicle:
  def __init__(self, vehID, wheels, make, model, color, engine):
   self.__vehType = "Vehicle"
   self.__vehID = vehID
   self.__wheels = wheels
   self.__make = make
   self.__model = model
   self.__color = color
   self.__engine = engine

  def set_vehType(self, vehType): self.__vehType = vehType
  def set_color(self, color): self.__color = color
  def set_engn_status(self, sts="Working"):
   self.__engine.set_estatus(sts)

  def get_vehType(self): return self.__vehType
  def get_engn(self): return self.__engine
  def get_engn_status(self): return self.__engine.get_estatus()
  def get_info(self):
   return f"{self.__color} {self.__wheels}-wheeler, {self.__make} {self.__model}.\nEngine:{self.__engine.get_engnID()}"

  def move(self, txt="The"):
   estatus = self.get_engn_status()
   if estatus != "Working":
    raise Exception(f"Couldn't start: Engine is {estatus}")
   else:
    print(f"{txt} {self.get_vehType} is moving.")
```

*SubClass*

```python
class Car(Vehicle):
  def __init__(self, vehID, wheels, make, model, color, engine):
    super().__init__(vehID, wheels, make, model, color, engine)
    self.set_vehType("Car")

  def move(self, txt="The"):
    # check status via the inherited getter
    engn = super().get_engn()
    current_status = engn.get_estatus()
    if current_status != "Working":
      raise Exception(f"The {self.get_vehType()} couldn't start: Engine is {current_status} ")
    else:
      print(f"{txt} {self.get_vehType()} is being driven.")
```

**Weak Relationship (Aggregation)**

A weak relationship (aggregation) is a relationship where objects are independent and can exist without each other.

🧠 Key Terms

Container class → use karta hai component

Component class → independent hota hai

💡 Roman Example

👉 Teacher aur Students

Teacher alag exist kar sakta hai

Student alag exist kar sakta hai

👉 Dono ek dusre pe dependent nahi

```python
# Example:
class Student:
    def __init__(self, name):
        self.name = name

class Teacher:
    def __init__(self, student):
        self.student = student  # weak relation
```

📌 Set to Problem → Set back to Working

👉 Problem:

Teacher remove ho jaye → student still exist

👉 Working:

Objects separately create hote hain

*Component Class*

```python
class AC:
  def __init__(self, acID):
   self.__acID = acID
   self.__estatus = "Working"

  def get_acstatus(self):
    return self.__estatus

  def set_acstatus(self, acstatus):
    self.__estatus = acstatus

  def get_acID(self):
    return self.__acID
```

*Container/Base Class*

```python
class Vehicle:
  def __init__(self, vehID, wheels, make, model, color, engine, ac=None):
    self.__vehType = "Vehicle"
    self.__vehID = vehID
    self.__wheels = wheels
    self.__make = make
    self.__model = model
    self.__color = color
    self.__engine = engine
    self.__ac = ac

  def set_vehType(self, vehType):
    self.__vehType = vehType

  def set_color(self, color):
    self.__color = color

  def set_ac(self, ac):
    self.__ac = ac

  def set_engn_status(self, sts="Working"):
    self.__engine.set_estatus(sts)

  def set_ac_status(self, sts="Working"):
    self.__ac.set_acstatus(sts)

  def get_vehType(self):
    return self.__vehType

  def get_engn(self):
    return self.__engine

  def get_engn_status(self):
    return self.__engine.get_estatus()

  def get_ac(self):
    return self.__ac

  def get_ac_status(self):
    return self.__ac.get_acstatus()

  def get_info(self):
    return f"{self.__color} {self.__wheels} wheeler, {self.__make} {self.__model}\nEngine: {self.__engine.get_engnID()}, {self.__ac.get_acID()}"

  def move(self, txt="The"):
    estatus = self.get_engn_status()
    if estatus != "Working":
      raise Exception(f"The {self.get_vehType()} couldn't start: Engine is {estatus} ")
    else:
      print(f"{txt} {self.get_vehType()} is moving.")
```

*SubClass*

```python
class Car(Vehicle):
  def __init__(self, vehID, wheels, make, model, color, engine, ac=None):
    # pass the actual ac variable, not None
    super().__init__(vehID, wheels, make, model, color, engine, ac)
    self.set_vehType("Car")

  def move(self, txt="The"):
    engn = super().get_engn()
    current_status = engn.get_estatus()
    if current_status != "Working":
      raise Exception(f"Couldn't start: Engine {current_status}")
    else:
      print(f"{txt} {self.get_vehType()} is being driven.")
```

**More Integral Component**

A more integral component is a highly dependent part of a system that plays a critical role in its functioning.

💡 Roman Example

👉 Human aur Heart

Heart = integral component

Heart ke bina system (body) kaam nahi karega

```python
class Heart:
    def beat(self):
        print("Heart is beating")

class Human:
    def __init__(self):
        self.heart = Heart()  # highly integral
```

📌 Set to Problem → Set back to Working

👉 Problem:

Heart remove → system fail

👉 Working:

Heart system ka core part hai

*Component Class*

```python
class Battery:
  def __init__(self, btryID):
    self.__btryID = btryID
    self.__btrystatus = "Working"

  def set_btrystatus(self, btrystatus):
    self.__btrystatus = btrystatus

  def get_btryID(self): return self.__btryID
  def get_btrystatus(self): return self.__btrystatus
```

*Base Class*

```python
class Vehicle:
  def __init__(self, vehID, wheels, make, model, color, engine, battery, ac=None):
    self.__vehType = "Vehicle"
    self.__vStatus = "Working"

    self.__vehID = vehID
    self.__wheels = wheels
    self.__make = make
    self.__model = model
    self.__color = color
    self.__engine = engine
    self.__battery = battery
    self.__ac = ac

  def set_vehType(self, vehType): self.__vehType = vehType
  def set_color(self, color): self.__color = color

  def set_ac(self, ac): self.__ac = ac

  def set_engn_status(self, sts="Working"):
    self.__engine.set_estatus(sts)
    if sts != "Working":
     self.__vStatus = sts
    else:
      if self.__battery.get_btrystatus() == "Working":
       self.__vStatus = sts

  def set_btry_status(self, sts="Working"):
    self.__battery.set_btrystatus(sts)
    if sts != "Working":
      self.__vStatus = sts
    else:
      if self.__engine.get_estatus() == "Working":
        self.__vStatus = sts

  def set_ac_status(self, sts="Working"):
    self.__ac.set_acstatus(sts)

  def get_vehID(self): return self.__vehID
  def get_vehType(self): return self.__vehType
  def get_vStatus(self): return self.__vStatus

  def get_engn(self): return self.__engine
  def get_engn_status(self): return self.__engine.get_estatus()

  def get_btry(self): return self.__battery
  def get_btry_status(self): return self.__battery.get_btrystatus()

  def get_ac(self): return self.__ac
  def get_ac_status(self): return self.__ac.get_acstatus()

  def get_info(self):
    return f"{self.__color} {self.__wheels}-wheeler, {self.__make} {self.__model}.\nEngine:{self.__engine.get_engnID()},Battery:{self.__battery.get_btryID()}"

  def move(self, txt="The"):
    current_status = self.get_vStatus()
    if current_status != "Working":
      raise Exception(f"The {self.get_vehType()} couldn't start: Status = Engine is {current_status}.")
    else:
      print(f"{txt} {self.get_vehType()} is moving.")
```

*SubClass*

```python
class Car(Vehicle):
  def __init__(self, vehID, wheels, make, model, color, engine, battery, ac=None):
    super().__init__(vehID, wheels, make, model, color, engine, battery, ac)
    self.set_vehType("Car")

  def move(self, txt="The"):
    current_status = super().get_vStatus()
    if current_status != "Working":
      raise Exception(f"The {self.get_vehType()} couldn't start: Status = Engine is {current_status}")
    else:
      print(f"{txt} {self.get_vehType()} is being driven.")
```
