```python
%%writefile mymodule.py

def Calc(num1, num2, Op):
    if Op == "+":
        res = num1 + num2
    elif Op == "-":
        res = num1 - num2
    elif Op == "*":
        res = num1 * num2
    elif Op == "/":
        res = num1 / num2
    elif Op == "//":
        res = num1 // num2
    elif Op == "**":
        res = num1 ** num2
    elif Op == "%":
        res = num1 % num2
    else:
        res = "N/A"
    return res

def Sum(Lst):
    Sm = 0
    for i in Lst:
        Sm = Sm + i
    return Sm

def Grade(Mrks):
    if Mrks >= 90:
        Grd = "A+"
    elif Mrks >= 85:
        Grd = "A"
    elif Mrks >= 80:
        Grd = "B+"
    elif Mrks >= 75:
        Grd = "B"
    elif Mrks >= 70:
        Grd = "C+"
    elif Mrks >= 65:
        Grd = "C"
    elif Mrks >= 60:
        Grd = "D"
    else:
        Grd = "F"
    return Grd

def TConv(T, Conv):
    if Conv in ("F", "f"):
        Res = T * 9/5 + 32
        unit = "C"
    elif Conv in ("C", "c"):
        Res = (T - 32) * 5/9
        unit = "F"
    else:
        Res = "N/A"; unit = ""
    return Res, unit

def tab(s=1, e=10, T=2):
    for i in range(s, e+1):
        print(f"{T} x {i} = {T * i}")

def fib(n):
    if n <= 1:
        res = n
    else:
        res = fib(n - 1) + fib(n - 2)
    return res
```
