4) **Word Frequency Visualization**

Load a text dataset and calculate the frequency distribution of words. Use Matplotlib or seaborn to create a bar chart of the top 10 most common words, nsuring you filter out noise such as punctuation and whitespace.

```python
import nltk
import string
from collections import Counter
import matplotlib.pyplot as plt
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

```python
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
```

```python
# Sample text dataset
text = """
Data Science is an interdisciplinary field.
Data analysis helps extract useful information from data.
Machine learning and data visualization are important in Data Science.
"""
```

```python
# Convert text to lowercase
text = text.lower()

# Tokenization
tokens = word_tokenize(text)

# Remove punctuation and stopwords
stop_words = set(stopwords.words('english'))

cleaned_words = []

for word in tokens:
    if word not in stop_words and word not in string.punctuation:
        cleaned_words.append(word)
```

```python
# Calculate word frequencies
word_freq = Counter(cleaned_words)
```

```python
# Top 10 most common words
top_words = word_freq.most_common(10)

# Separate words and counts
words = [word for word, count in top_words]
counts = [count for word, count in top_words]

# Create bar chart
plt.figure(figsize=(10,5))
plt.bar(words, counts)

# Labels and title
plt.xlabel("Words")
plt.ylabel("Frequency")
plt.title("Top 10 Most Common Words")

# Show graph
plt.show()
```

5) **Loss Function Tracking**

Write a python script to calculate a binary cross entropy loss for a set of predicted values and true labels. Generate a line chart to visualize how the loss changes over 10 simulated training epochs.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# True labels
y_true = np.array([1, 0, 1, 1, 0])
```

```python
# Simulated predicted probabilities for 10 epochs
predictions = [
    [0.50, 0.50, 0.50, 0.50, 0.50],
    [0.60, 0.40, 0.55, 0.58, 0.45],
    [0.65, 0.35, 0.60, 0.62, 0.40],
    [0.70, 0.30, 0.68, 0.66, 0.35],
    [0.75, 0.25, 0.72, 0.70, 0.30],
    [0.80, 0.20, 0.78, 0.75, 0.25],
    [0.84, 0.16, 0.82, 0.80, 0.20],
    [0.88, 0.12, 0.86, 0.84, 0.15],
    [0.92, 0.08, 0.90, 0.88, 0.10],
    [0.95, 0.05, 0.94, 0.92, 0.05]
]
```

```python
# Function to calculate Binary Cross Entropy Loss
def binary_cross_entropy(y_true, y_pred):

    y_pred = np.clip(y_pred, 1e-10, 1-1e-10)

    loss = -np.mean(
        y_true * np.log(y_pred) +
        (1 - y_true) * np.log(1 - y_pred)
    )

    return loss
```

```python
# Store loss values
loss_values = []
```

```python
# Calculate loss for each epoch
for pred in predictions:
    loss = binary_cross_entropy(y_true, np.array(pred))
    loss_values.append(loss)

# Epoch numbers
epochs = range(1, 11)
```

```python
# Plot Loss Curve
plt.figure(figsize=(8,5))
plt.plot(epochs, loss_values, marker='o')

plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Binary Cross Entropy Loss over Epochs")

plt.grid(True)
plt.show()
```

```python
# Print loss values
print("Loss Values:")
print(loss_values)
```
