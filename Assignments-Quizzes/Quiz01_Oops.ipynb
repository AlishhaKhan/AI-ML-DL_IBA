Situation: You’re building an e-commerce system with the following requirements: • Products can be physical items or digital downloads • All products have name, price, and description • Physical products have weight and shipping cost • Digital products have file size and download link • Customers can add products to cart and calculate total cost • Digital products have no shipping cost Task: Design the class hierarchy and implement: 1. Base Product class 2. PhysicalProduct and DigitalProduct subclasses 3. ShoppingCart class that can handle both types 4. Method to calculate total cost including shipping



```
# Base Class
class Product:
    def __init__(self, name, price, description):
        self.name = name
        self.price = price
        self.description = description

    def get_price(self):
        return self.price


# Physical Product
class PhysicalProduct(Product):
    def __init__(self, name, price, description, weight, shipping_cost):
        super().__init__(name, price, description)
        self.weight = weight
        self.shipping_cost = shipping_cost

    def get_price(self):
        return self.price + self.shipping_cost


# Digital Product
class DigitalProduct(Product):
    def __init__(self, name, price, description, file_size, download_link):
        super().__init__(name, price, description)
        self.file_size = file_size
        self.download_link = download_link

    def get_price(self):
        return self.price   # no shipping


# Shopping Cart
class ShoppingCart:
    def __init__(self):
        self.items = []

    def add_product(self, product):
        self.items.append(product)

    def total_cost(self):
        total = 0
        for item in self.items:
            total += item.get_price()
        return total


# Example
cart = ShoppingCart()

p1 = PhysicalProduct("Laptop", 1000, "Gaming laptop", 2, 50)
p2 = DigitalProduct("Ebook", 100, "Python book", "5MB", "link")

cart.add_product(p1)
cart.add_product(p2)

print(cart.total_cost())
```
