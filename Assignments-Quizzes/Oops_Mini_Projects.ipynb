**1) Bank Account**

* Simulate:

* Create Account

* Deposite

* Withdraw

* Check balance

```python
class BankAccount:
  def __init__(self, name, balance=0):
    self.name = name
    self.balance = balance


  def deposite(self, amount):
   self.balance += amount
   print(f"{amount} deposited. New balance: {self.balance}")


  def withdraw(self, amount):
    if amount > self.balance:
     print(f"Insufficient balance")
    else:
     self.balance -= amount
     print(f"{amount} withdrawn.  Remaining balance: {self.balance}")

  def check_balance(self):
   print(f"Amount Holder: {self.name}, Balance: {self.balance}")
```

```python
acc1 = BankAccount("Alisha", 1000)
acc1.deposite(500)
acc1.withdraw(200)
acc1.check_balance()
```

**2) University Management**

Simulation:

Add students

Assign subjects

Add marks

Show result

```python
class Student:
  def __init__(self, name, roll_no):
    self.name = name
    self.roll_no = roll_no
    self.marks = {}


  def add_marks(self, subjects, marks):
    self.marks[subjects] = marks

  def display_result(self):
    print(f"\n Students: {self.name} | Roll No: {self.roll_no}")

    total = 0
    for subject, marks in self.marks.items():
     print(f"{subject}: {marks}")
     total += marks


    if len(self.marks) > 0:
      avg = total / len(self.marks)
      print(f"Average: {avg}")
    else:
      print("No subjects added")


class University:
  def __init__(self, name):
    self.name = name
    self.students = []


  def add_students(self, students):
   self.students.append(students)

  def show_all_students(self):
    print(f"\nUniversity: {self.name}")
    for students in self.students:
      students.display_result()
```

```python
s1 = Student("Alisha", "101")
s1.add_marks("Math", 85)
s1.add_marks("Stats", 90)

s2 = Student("Sara", "102")
s2.add_marks("Math", 70)
s2.add_marks("Stats", 75)

uni = University("IBA")

uni.add_students(s1)
uni.add_students(s2)

uni.show_all_students()
```
