1. Sum of List

 Given a list of numbers, use a for loop to calculate the total sum.

```python
numbers = [10, 20, 30, 40, 50]

total_sum = 0
for num in numbers:
    total_sum += num

print("Total Sum:", total_sum)
```

2. Count Even Numbers

Given a list of integers, count how many numbers are even.

```python
numbers = [1, 2, 3, 4, 5, 6, 8]

count = 0
for num in numbers:
    if num % 2 == 0:
        count += 1

print("Even Numbers Count:", count)
```

3. Find Maximum

 Write a program to find the largest number in a list (without using max()).

```python
numbers = [12, 45, 7, 89, 23]

largest = numbers[0]

for num in numbers:
    if num > largest:
        largest = num

print("Largest Number:", largest)
```

5. Count Positive, Negative, Zero Given a list, count how many positive, negative, and zero values exist.

```python
numbers = [10, -5, 0, 20, -8, 0, 3]

positive = 0
negative = 0
zero = 0

for num in numbers:
    if num > 0:
        positive += 1
    elif num < 0:
        negative += 1
    else:
        zero += 1

print("Positive:", positive)
print("Negative:", negative)
print("Zero:", zero)
```

6. Reverse a List (Manually) Create a new list that contains elements of the original list in reverse order (without using reverse()).

```python
numbers = [1, 2, 3, 4, 5]

reversed_list = []

for i in range(len(numbers) - 1, -1, -1):
    reversed_list.append(numbers[i])

print("Reversed List:", reversed_list)
```

7. Remove Duplicates Given a list, create a new list that contains only unique elements.

```python
numbers = [1, 2, 2, 3, 4, 4, 5]

unique_list = []

for num in numbers:
    if num not in unique_list:
        unique_list.append(num)

print("Unique Elements:", unique_list)
```

8. Find Second Largest Number Find the second largest number in a list.

```python
numbers = [10, 20, 4, 45, 99]

largest = second_largest = float('-inf')

for num in numbers:
    if num > largest:
        second_largest = largest
        largest = num
    elif num > second_largest and num != largest:
        second_largest = num

print("Second Largest:", second_largest)
```

9. Check Palindrome List Check if a list reads the same forward and backward. Example: [1, 2, 3, 2, 1] = Palindrome

```python
numbers = [1, 2, 3, 2, 1]

is_palindrome = True

for i in range(len(numbers) // 2):
    if numbers[i] != numbers[len(numbers) - i - 1]:
        is_palindrome = False
        break

if is_palindrome:
    print("Palindrome")
else:
    print("Not Palindrome")
```

10. Count Occurrence of Element Given a list and a number, count how many times that number appears.

```python
numbers = [1, 2, 3, 2, 4, 2, 5]
target = 2

count = 0

for num in numbers:
    if num == target:
        count += 1

print("Occurrence of", target, ":", count)
```

11. FizzBuzz
Print numbers from 1 to 50:

- If divisible by 5 then print 'Buzz'

13. Separate Even and Odd Lists Given a list, create two separate lists: - One for even numbers - One for odd numbers

```python
numbers = [1, 2, 3, 4, 5, 6, 7, 8]

even_list = []
odd_list = []

for num in numbers:
    if num % 2 == 0:
        even_list.append(num)
    else:
        odd_list.append(num)

print("Even Numbers:", even_list)
print("Odd Numbers:", odd_list)
```

14. Find Common Elements Given two lists, print elements that appear in both lists.

```python
list1 = [1, 2, 3, 4, 5]
list2 = [3, 4, 5, 6, 7]

common = []

for num in list1:
    if num in list2 and num not in common:
        common.append(num)

print("Common Elements:", common)
```

15. Check Sorted List Check if a list is sorted in ascending order.

```python
numbers = [1, 2, 3, 4, 5]

is_sorted = True

for i in range(len(numbers) - 1):
    if numbers[i] > numbers[i + 1]:
        is_sorted = False
        break

if is_sorted:
    print("List is sorted")
else:
    print("List is not sorted")
```

 16. Rotate List Left Rotate a list to the left by 1 position.
Example: [1,2,3,4] converted to [2,3,4,1]

```python
numbers = [1, 2, 3, 4]

rotated = []

for i in range(1, len(numbers)):
    rotated.append(numbers[i])

rotated.append(numbers[0])

print("Rotated List:", rotated)
```

19. Find All Prime Numbers in a List Given a list of integers, print only the prime numbers.

```python
numbers = [2, 3, 4, 5, 6, 7, 8, 9, 11]

print("Prime Numbers:")

for num in numbers:
    if num > 1:
        is_prime = True
        for i in range(2, num):
            if num % i == 0:
                is_prime = False
                break
        if is_prime:
            print(num)
```

20. Mini Student Grading System Given a list of marks: - 90+ is Grade A - 75–89 is Grade B - 50–74 is Grade C - Below 50 is Fail Print each student's grade using a loop.

```python
marks = [95, 82, 67, 45, 74, 88, 91]

for mark in marks:
    if mark >= 90:
        grade = "A"
    elif mark >= 75:
        grade = "B"
    elif mark >= 50:
        grade = "C"
    else:
        grade = "Fail"

    print("Marks:", mark, "Grade:", grade)
```
