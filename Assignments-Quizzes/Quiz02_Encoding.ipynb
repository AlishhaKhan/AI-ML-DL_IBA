```python
# handle data
import pandas as pd
import numpy as np
```

```python
# visualization
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load dataset
df = pd.read_csv('/content/PlayGolfData.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
print(df['Temperature'].unique())
print(df['Humidity'].unique())
```

```python
from sklearn.preprocessing import LabelEncoder     # Label encoding
le = LabelEncoder()
df['Wind'] = le.fit_transform(df['Wind'])
```

```python
df = pd.get_dummies(df, columns=['Outlook']) # one-hot encoding
```

```python
from sklearn.preprocessing import OrdinalEncoder     # ordinal encoding
oe = OrdinalEncoder(categories=[['Low', 'Medium', 'High', 'Extreme']])
df['Temperature'] = oe.fit_transform(df[['Temperature']])
```

```python
print(df['Temperature'].unique())
print(df['Humidity'].unique())
```

```python
print(df['Temperature'].value_counts())
```

```python
sns.histplot(df['Temperature'])
plt.show()
```
